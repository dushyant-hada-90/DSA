Generate Binary Strings Without Consecutive 1s

```python
class Solution:
    def generateBinaryStrings(self, n):
        output=[]
        def backtrack(idx,prev,curr):
            # print("---------",idx,prev,curr,"---------")
            if idx==n:
                output.append(curr)
                # print("returning...",output)
                return
            backtrack(idx+1,0,curr+"0")

            if prev!=1:
                backtrack(idx+1,1,curr+"1")
        backtrack(0,0,"")   
        return output     

# count is fib(n)
# o(fib(n)),o(n) recursion stack
```

```python
sol=Solution()
sol.generateBinaryStrings(3)
```

Generate Parentheses



```python
class Solution:
    def generateParenthesis(self, n):
        output=[]
        def backtrack(opened,closed,curr):
            if closed == n:
                output.append(curr)
                return
            if opened<n:
                backtrack(opened+1,closed,curr+"(")
            if opened>closed:
                backtrack(opened,closed+1,curr+")")

        backtrack(0,0,"")   
        return output     

# o(4^n/((n^1.5)*(pi*0.5))),o(n) recursion stack , o(C_n*n) is output storage
```

```python
sol=Solution()
sol.generateParenthesis(1)
```

Subsets

```python
class Solution:
    def subsets(self,nums):
        output=[]
        n=len(nums)
        def backtrack(idx,curr):
            # print("---------",idx,curr,"---------")
            if idx==n:
                output.append(curr[:])
                # print("returning...",output)
                return
            # print(curr)
            backtrack(idx+1,curr)
            curr.append(nums[idx])
            backtrack(idx+1,curr)
            curr.pop()
            # print(curr,"popped")
        backtrack(0,[])   
        return output     

# o(2^n*n),o(n) recursion stack,o(n*2^n) for output storage
```

```python
sol=Solution()
sol.subsets([1,2,3])
```

Count all subsequences with sum K




```python
class Solution:
    def countSubsequenceWithTargetSum(self, nums, k):
        count=0

        def backtrack(idx,s):

            nonlocal count
            if idx>=len(nums):
                if s==k :
                    count+=1
                return 
            backtrack(idx+1,s)

            backtrack(idx+1,s+nums[idx])
           
            
        backtrack(0,0)
        return count
    
# o(2^n),o(1) , o(1) for output returned
```

Check if there exists a subsequence with sum K




```python
class Solution:
    def checkSubsequenceSum(self, nums, k):

        def check(idx,s):
            if idx>=len(nums):
                if s==k:
                    return True
                return False
            

            if (check(idx+1,s)):
                return True
      
            
            if check(idx+1,s+nums[idx]):
                return True
                
            return False
        

        return check(0,0)
```

Combination Sum

```python
class Solution:
    def combinationSum(self, candidates, target):
        output = []

        def backtrack(idx, s, curr, depth=0):
            print("  " * depth, idx, s, curr)

            if s == target:
                output.append(curr[:])
                print("  " * depth, "sum found")
                return
            if idx >= len(candidates):
                print("  " * depth, "out of index")
                return

            # pick
            if s + candidates[idx] <= target:
                print("  " * depth, "picking...")
                curr.append(candidates[idx])
                backtrack(idx, s + candidates[idx], curr, depth + 1)
                curr.pop()

            # not pick
            print("  " * depth, "not picking")
            backtrack(idx + 1, s, curr, depth + 1)

            print("  " * depth, "over")

        backtrack(0, 0, [])
        return output
```

```python
sol=Solution()
sol.combinationSum([2,3,6,7],7)
```

```python
class Solution:
    def combinationSum2(self, candidates, target):
        output = []
        candidates.sort() #nlogn

        def backtrack(idx=0,s=0,curr=[],space=1): #2^n if all subsets add upto target
            
            print(space*"  ",idx,s,curr)

            if s == target:
                output.append(curr[:])  #k*n if there are k valid solutions
                return
            if idx>=len(candidates):
                return
            
            for i in range(idx,len(candidates)):
                print(space*"  ","loop started",i)
                if i>idx and  candidates[i]==candidates[i-1]:
                    print(space*"  ","encountered same element again in same depth")
                    continue

                # pick
                curr.append(candidates[i])
                backtrack(i+1,s+candidates[i],curr,space+1)
                curr.pop()
                
                if s+candidates[i]>target:
                    break #since  they are soted , no need to move further
             
                
            
        
        backtrack()
        return output
    
# O(2^n)+O(nlogn)+O(kâ‹…n), o(n) is recursion depth , o(n) is curr ,o(k*n) is output returned if there are k valid combinations
```

```python
sol=Solution()
sol.combinationSum2([1,1,2], target = 2)
```

```python
class Solution:
    def subsetSums(self, nums):
        output = [] 
        def backtrack(idx=0,s=0): #2^n for exploring all possibilities
            if idx>=len(nums):
                s not in output and output.append(s) #o(n)
                return
            # pick
            backtrack(idx+1,s+nums[idx])
            # not pick
            backtrack(idx+1,s)
        
        backtrack()
        return list(set(output))

#o(n*2^n) , o(2^n) for returning , o(n) recursion depth
```

```python
sol=Solution()
sol.subsetSums([3,2,5])
```

Subsets II

```python
class Solution:
    def subsetsWithDup(self, nums):
        output= []
        nums.sort()
        def backtrack(idx=0,curr=[]):

            output.append(curr[:])
        
            
            for i in range(idx,len(nums)):
                if i>idx and nums[i]==nums[i-1]:
                    continue #do not explore to avoid duplicacy
                # pick
                curr.append(nums[i])
                backtrack(i+1,curr)
                curr.pop()
                
        backtrack()
        return output
    # o(2^n+nlogn),o(2^n) for returning , o(n) recursion depth
```

```python
class Solution:
    def subsetsWithDup(self, nums):
        output= []
        nums.sort()
        def backtrack(idx=0,curr=[]):
            if idx>=len(nums):
                output.append(curr[:])  #o(n)
                return

            # pick
            curr.append(nums[idx])
            backtrack(idx+1,curr)
            curr.pop()

            # not pick
            while idx+1<len(nums) and nums[idx]==nums[idx+1]:
                idx+=1
            backtrack(idx+1,curr)
                
        backtrack()
        return output
    # o(n2^n+nlogn),o(n*2^n) for returning , o(n) recursion depth
```

```python
sol=Solution()
sol.subsetsWithDup([1,2,3])
```

Combination Sum III

```python
class Solution:
    def combinationSum3(self, k: int, n: int):
        output = []
        def backtrack(s=0,curr=[],number=1):
            if len(curr)==k:
                if s==n:    
                    output.append(curr[:]) #o(n)
                return
            if number>9 or s>n:
                return
            
            # pick a number
            curr.append(number)
            backtrack(s+number,curr,number+1)
            curr.pop()

            # don't pick a number
            backtrack(s,curr,number+1)
        backtrack()
        return output
# o(9Ck*n),o(k*9Ck) spac for returning,o(9) stack space
            
```

```python
sol=Solution()
sol.combinationSum3(3,9)
```

Letter Combinations of a Phone Number

```python
class Solution:
    def letterCombinations(self, digits: str):
        if not digits:
            return []
        output=[]
        phone_map = {
            "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
            "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
        }
        def backtrack(idx=0,curr=""):
            if idx==len(digits):
                output.append(curr) #o(n) because python by default copies the strings before storing
                return
            letters=phone_map[digits[idx]]
            for letter in letters:
                backtrack(idx+1,curr+letter)
        
        backtrack()
        return output
# n=len(digits)
# o((4^n or 3^n)*n) , o(4^n) for output,o(n) stack space
        
```

```python
sol=Solution()
sol.letterCombinations("")
```
