Palindrome Partitioning

```python
class Solution:
    def partition(self, s: str) :
        output = []
        def backtrack(idx=0,curr=[],space=0): #o(2^n)
            print("  "*space,idx,curr)
            if idx==len(s):
                output.append(curr[:]) #o(n)
                return
            for i in range(idx,len(s)):
                sub = s[idx:i+1]
                print(space*"  ",sub)
                if sub==sub[::-1]:  #o(n)
                    curr.append(sub)
                    backtrack(i+1,curr,space+1)
                    curr.pop()

        backtrack()
        return output
# there are 2^(n-1) ways to partition
# o(n*2^n),o(n) stack space,o(n*2^n) for output
```

Word Search

```python
class Solution:
    def exist(self, board,word):

        m=len(board)
        n=len(board[0])

        def backtrack(i,j,k,space=0):
            print(space*"  ",i,j,word[:k])
            if board[i][j]!=word[k]:
                return False
                
            if k==len(word)-1:
                return True
            
            # mark visited
            temp=board[i][j]
            board[i][j]="#"

            # exploration 
            for di,dj in [(1,0),(-1,0),(0,1),(0,-1)]:
                ni,nj = i+di,j+dj
                if 0<=ni<m and 0<=nj<n and board[ni][nj]!="#":
                    if backtrack(ni,nj,k+1):
                        board[i][j]=temp
                        return True



            # restore
            board[i][j]=temp
            return False
        # try every cell as starting point
        for i in range(m):
            for j in range(n):
               if backtrack(i,j,0):
                   return True
        return False
                    
            
        
```

```python
board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "ABCCED"
sol=Solution()
sol.exist(board,word)
```
