String to Integer (atoi)

```python
class Solution:
    def myAtoi(self, s: str) -> int:
        num=0
        sign=1
        started = False
        for i in s:
            if i==" " and not started:
                continue
            elif i=="-" and not started:
                sign = -1
                started = True
            
            elif 48<=ord(i)<=57:
                num=num*10+int(i)
                started = True
            else:
                break
        return num*sign
            
            
```

Pow(x, n)

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        if n==0:
            return 1

        if n<0:
            x=1/x
            n=-n

        if n%2==0:
            return self.myPow(x*x,n//2)
        else:
            return x*self.myPow(x*x,(n-1)//2)
# o(logn base 2),o(logn base 2) as each power calls its function
```

```python
sol=Solution()
sol.myPow(2,5)
```

Count Good Numbers

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        mod=10**9+7
        curr = x%mod
        result = 1
        while n>0:
            # print(f'curr:{curr},result:{result},n:{n}')
            if n %2!= 0: 
                result = (result*curr)%mod
            curr = curr*curr%mod
            n=n//2  
        return result

        
    def countGoodNumbers(self, n: int) -> int:
        mod=10**9+7
        return self.myPow(5,(n+1)//2)*self.myPow(4,n//2)%mod
# o(logn base 2),o(1)
```

```python
sol=Solution()
sol.countGoodNumbers(21)
```

Reverse a Stack




```python
class Solution:
    def insertAtBottom(self,stack,item):
        # print(stack)
        if len(stack)==0:
            stack.append(item)
            return
        top=stack.pop()
        self.insertAtBottom(stack,item)
        stack.append(top)

    def reverseStack(self, stack):
        if len(stack)==0:
            return stack
        curr = stack.pop()
        self.reverseStack(stack)
        self.insertAtBottom(stack,curr)
# o(n**2),o(n)
```

```python
stack=[1,2,3,4]
rev = Solution()
rev.reverseStack(stack)
print(stack)
```

sort a stack

```python
class Solution:
    def insertInSortedStack(self,stack,item):
        if not stack or item>=stack[-1]:
            stack.append(item)
            return
        top = stack.pop()
        self.insertInSortedStack(stack,item)
        stack.append(top)
    
    def sortStack(self,stack):
        if not stack:
            return
        top = stack.pop()
        self.sortStack(stack)
        self.insertInSortedStack(stack,top)
    # o(n**2),o(n)
```

```python
stack = [3, 1, 4, 2, 5]
sol=Solution()
sol.sortStack(stack)
print(stack)  
```
