Topo Sort

```python
class Solution:
    def topoSort(self,graph):
        stack = []
        visited = set()

        def dfs(node):
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    dfs(neighbor)
            stack.append(node)

        for i in range(len(graph)):
            print(i,stack)
            if i not in visited:
                dfs(i)

                
        print(stack[::-1])
# o(V+E) for dfs
# o(v) for stack space , o(v) for recursion stack
sol = Solution()
graph = [[],[],[3],[1],[0,1],[0,2]]
sol.topoSort(graph)
                    
```

Kahn's Algorithm

```python
from collections import deque
class Solution:
    def kahnAlgo(self,graph):
        q = deque([])
        indeg = [0]*len(graph)
        ans = []

        for i in range(len(graph)):
            for j in graph[i]:
                indeg[j]+=1

        for i in range(len(indeg)):
            if indeg[i] == 0:
                q.append(i)
                ans.append(i)

        while q:
            print(indeg,q)  
            node = q.popleft()
            for neighbor in graph[node]:
                indeg[neighbor]-=1
                if indeg[neighbor] == 0:
                    ans.append(neighbor)
                    q.append(neighbor)
        return ans

# o(V+E)  for bfs
# o(V)
sol = Solution()
graph = [[],[],[3],[1],[0,1],[0,2]]
sol.kahnAlgo(graph)
```

Course Schedule

```python
class Solution:
    def canFinish(self, numCourses, prerequisites):
        # first we will express prerequisites as a directed graph and then perform topoSort , if it fails then that means that the graph is cyclic and it is impossible to complete all courses'
        graph = [[] for _ in range(numCourses)]
        for i in range(len(prerequisites)):
            graph[prerequisites[i][1]].append(prerequisites[i][0])
                
        q = deque([])
        ans = []
        indeg = [0]*len(graph)
        for i in range(len(graph)):
            for j in graph[i]:
                indeg[j]+=1
        for i in range(len(indeg)):
            if indeg[i] == 0:
                q.append(i)
                ans.append(i)
        while q:
            node = q.popleft()
            for neighbor in graph[node]:
                indeg[neighbor] -= 1
                if indeg[neighbor] == 0:
                    q.append(neighbor)
                    ans.append(neighbor)
        return len(ans) == numCourses
# o(V+E)  for bfs
# o(V)
sol = Solution()
sol.canFinish(numCourses = 2, prerequisites = [[1,0],[0,1]])
```

Course Schedule II

```python
class Solution:
    def findOrder(self, numCourses, prerequisites):
        # first we will express prerequisites as a directed graph and then perform topoSort , if it fails then that means that the graph is cyclic and it is impossible to complete all courses'
        graph = [[] for _ in range(numCourses)]
        for i in range(len(prerequisites)):
            graph[prerequisites[i][1]].append(prerequisites[i][0])
                
        q = deque([])
        ans = []
        indeg = [0]*len(graph)
        for i in range(len(graph)):
            for j in graph[i]:
                indeg[j]+=1
        for i in range(len(indeg)):
            if indeg[i] == 0:
                q.append(i)
                ans.append(i)
        while q:
            node = q.popleft()
            for neighbor in graph[node]:
                indeg[neighbor] -= 1
                if indeg[neighbor] == 0:
                    q.append(neighbor)
                    ans.append(neighbor)
        return ans if  len(ans) == numCourses else []
# o(V+E)  for bfs
# o(V)
sol = Solution()
sol.findOrder(numCourses = 2, prerequisites = [[1,0],[0,1]])
```

Find eventual safe states

```python
# if u wish to use topoSort then first of all reverse the entire graph and then do a toposort , the ans array willl be returned
```

```python
class Solution:
    # A node is safe if all paths starting from it eventually reach a terminal node (a node with no outgoing edges).
    def eventualSafeNodes(self, graph):
        n = len(graph)
        visited = [0]*n
        pathVis = [0]*n
        safe = [0]*n

        def dfs(node):
            visited[node] = 1
            pathVis[node] = 1
            for neighbor in graph[node]:
                if not visited[neighbor]:
                    if dfs(neighbor) == False:
                        return False
                elif pathVis[neighbor]: #means that the node is visited and in the same stack_path , implying that we have reached that node again -> cycle detected -> unsafe path
                    return False
            pathVis[node] = 0 #after a node has been processed remove it from stack_path by marking it as 0
            safe[node] = 1 
            return True 
             
        for i in range(len(graph)):
            if not visited[i]:
                print(i,"---")
                dfs(i)
            
        return [i for i in range(n) if safe[i] == 1]
sol = Solution()
sol.eventualSafeNodes(graph = [[1,2],[2,3],[5],[0],[5],[],[]])
```

Alien Dictionary

```python
from collections import defaultdict,deque
class Solution:
    def alienDict(self,words):
        chars = set(''.join(words))
        graph = {ch:[] for ch in chars}
        for i in range(len(words)-1):
            s1 = words[i]
            s2 = words[i+1]
            for j in range(min( len(s1),len(s2) )):
                if s1[j]!=s2[j]:
                    graph[s1[j]].append(s2[j])
                    break
        # now perform a topoSort to get the order about what comes before what in alien dictionary
        # note the graph may look visually cyclic but it is not.

        # create indeg
        indeg = {ch: 0 for ch in chars}
        for i in graph:
            for j in graph[i]:
                indeg[j] = indeg.get(j,0) + 1
        

        # start topoSort
        ans = []
        q = deque([])
        for i in indeg:
            if indeg[i] == 0:
                q.append(i)
                ans.append(i)

        while q:
            print(q)
            node = q.popleft()
            for neighbor in graph[node]:
                    indeg[neighbor] -= 1
                    if indeg[neighbor] == 0:
                        q.append(neighbor)
                        ans.append(neighbor)

        return "".join(ans)
# o(N + L) where n is len(words) and l is average word length
# o(C + E) where c is the number of unique characters and E is the number of  edges
sol = Solution()
sol.alienDict(['baa','abcd','abca','cab','cad'])
```
