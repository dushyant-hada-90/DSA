Tarjanâ€™s Algorithm /bridges

```python
# a bridge is just an edge on whose removal tge entire graph can be broken into multiple pieces
from collections import defaultdict
class Solution:
    def bridge(self,n,connections):
        graph = defaultdict(list)
        for u,v in connections:
            graph[u].append(v)
            graph[v].append(u)

        time_input = 0
        time,lowest_time = [0]*n,[0]*n
        visited = set()
        ans = []
        def dfs(node,parent):
            print("-"*node,node)
            nonlocal time_input
            time[node] = time_input
            lowest_time[node] = time_input
            print("--"*node,time,lowest_time,"---")
            time_input+=1
            visited.add(node)
            for nei in graph[node]:
                if nei not in visited:
                    lowest_time_nei = dfs(nei,node)
                    lowest_time[node] = min(lowest_time_nei,lowest_time[node])
                    # is there a bridge between us?
                    if lowest_time_nei>time[node]:
                        print("bridge between",node,nei)
                        ans.append([node,nei])
                if nei in visited and nei != parent:
                    print(node*"-",nei,"***")
                    lowest_time[node] = min(lowest_time[node],lowest_time[nei])
            print("-"*node,time,lowest_time)
            return lowest_time[node]
            
        dfs(0,-1)
        return ans
# o(V+E) for dfs
# o(V) 
Solution().bridge(4, [[0,1],[1,2],[2,0],[1,3]])
```

Articulation point

```python
# a artic..point is just a node on whose removal the remaining entire graph can be broken into multiple pieces
from collections import defaultdict
class Solution:
    def bridge(self,n,connections):
        graph = defaultdict(list)
        for u,v in connections:
            graph[u].append(v)
            graph[v].append(u)
        print(graph)

        time_input = 0
        time,lowest_time = [0]*n,[0]*n
        visited = set()
        ans = set()

        def dfs(node,parent):
            print("-"*node,node)
            nonlocal time_input
            time[node] = time_input
            lowest_time[node] = time_input
            print("--"*node,time,lowest_time,"---")
            time_input+=1
            visited.add(node) 
            child = 0

            for nei in graph[node]:
                if nei==parent:
                    continue
                if nei not in visited:
                    dfs(nei,node)
                    lowest_time[node] = min(lowest_time[nei],lowest_time[node])

                    # am i the arti.. point?
                    if lowest_time[nei]>=time[node] and parent!=-1:
                        print("articulation point",node)
                        ans.add(node)
                    child+=1
                else:
                    lowest_time[node] = min(lowest_time[nei],lowest_time[node])
            if parent == -1 and child>1:
                ans.add(node)


                
            print("-"*node,time,lowest_time)
            print("-"*node,ans)
            return lowest_time[node]
            
        dfs(0,-1)
        return ans
# o(V+E) for dfs
# o(V) 
Solution().bridge(4, [[0,1],[1,2],[2,0],[1,3]])
```
