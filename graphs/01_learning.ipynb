```python
from collections import deque

class Graph:
    def __init__(self,directed = False):
        self.graph = {}
        self.directed = directed
    
    def add_node(self,node):
        if node not in self.graph:
            self.graph[node] = []

    def add_edge(self,u,v):
        self.add_node(u) # if does not exist
        self.add_node(v) # if does not existt
        self.graph[u].append(v)  #graph mein key 'u' ko search karo toh as value tumhe ek list milegi about u kis kis se connected hai , uss list mein v bhi append kar do
        if not self.directed:
            self.graph[v].append(u)
    
    def remove_edge(self,u,v):
        if u in self.graph and v in self.graph[u]:
            self.graph[u].remove(v)
        if not self.directed:
            if v in self.graph and u in self.graph[v]:
                self.graph[v].remove(u)
    
    def remove_node(self,node):
        if node in self.graph:
            # remove its edges from other nodes
            for key in list(self.graph):
                if node in self.graph[key]:
                    self.graph[key].remove(node)
            # remove the node itself
            del self.graph[node]
    




    # helper for dfs
    def _dfs_util(self,node,visited,result):
        visited.add(node)
        result.append(node)
        for neighbor in self.graph.get(node,[]):
            if neighbor not in visited:
                self._dfs_util(neighbor,visited,result)
    
    # DFS for the whole graph (handles disconnected graphs)
    def dfs(self,start):
        visited = set() #so that we don't get lost by traversing same node again and again
        result = []

        for node in self.graph: #teaverse for each key in graph
            if node not in visited: #check if we had already been there
                self._dfs_util(node,visited,result) #now traverse each neighbor for that node
        return result
    # DFS visits each node once, each edge once: so o(node_count+edge_count),o(n) for set





    def dfs_iterative(self,start):
        visited = set()
        stack = [start]
        result = []
 
        while stack:
            node = stack.pop()
            if node not in visited:
                visited.add(node)
                result.append(node)
                # the order of visited nodes may differ in this iterative version from recursive one 
                for neighbor in self.graph.get(node,[]):
                    if neighbor not in visited:
                        stack.append(neighbor)

        return result





    def bfs(self,start):
        visited = set()
        q = deque([start])
        result = []
        visited.add(start)

        while q:
            node = q.popleft()
            result.append(node)
            for neighbor in self.graph.get(node,[]):
                if neighbor not in visited:
                    visited.add(neighbor)
                    q.append(neighbor)
        return result
    # o(node_count + edge_count)
    # o(node) for visited set
# Example usage:
g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)

print("DFS starting from node 1:",g.dfs(1))
print("BFS starting from node 1:",g.bfs(1))
```
