Subsequence with k sum

```python
class Solution:
    def func(self,arr,k):
        n = len(arr)

        dp = [[float("-inf")]*(k+1) for _ in range(n)]
        def helper(idx,target):
                if dp[idx][target]!=float("-inf"):
                     return dp[idx][target]
                if target==0:
                    return True
                if idx<0:
                     return False
                # either take it
                # only if it is smaller than target 
                take = False  #you need to define it before 
                if arr[idx]<=target:
                    take =  helper(idx-1,target-arr[idx])
                # or dont take it
                not_take = helper(idx-1,target)
                dp[idx][target] = take or not_take 
                return dp[idx][target]

        return helper(n-1,k)
# o(n*k) since there are n elements and max value of target is k so total n*k states are possible # time comp would have been 2^n without dp
# o(n*k) for dp 
# o(n) stack space as at each level we handle a index 
Solution().func([2,3,1,1],4)
```

Partition Set Into 2 Subsets With Min Absolute Sum Diff 

```python
import itertools
class Solution:
    def minimumDifference(self, nums):
        n = len(nums) // 2
        total_sum = sum(nums)

        # split into two halves
        left_part = nums[:n]
        right_part = nums[n:]

        # Helper to generate all subset sums grouped by the number of elements (k)
        # Returns a dictionary: { count_of_elements: [list_of_sums] }
        def get_sums(arr):
            sums_by_count = {}
            # iterate through all lengths from 0 to N
            for k in range(len(arr)+1):
                sums_by_count[k] = []
                for seq in itertools.combinations(arr,k):#generates all unique subsequences of k items from arr 
                    sums_by_count[k].append(sum(seq))
            return sums_by_count
        
        left_sums = get_sums(left_part)
        right_sums = get_sums(right_part)

        # sort right_sums for binary search
        for k in right_sums:
            right_sums[k].sort()
        
        min_diff = float("inf")
        half_total = total_sum/2 #we want out subset to be as close as possible to half_total as possible

        for k in range(n+1):
            left_k_sums = left_sums[k]
            # If we took k from left, we need n-k from right
            right_target_count = n-k
            right_k_sums = right_sums[right_target_count]

            # for each sum in the left group, find the best match in the right group
            for s_left in left_k_sums:
                # We want: s_left + s_right ≈ total_sum / 2
                # So: s_right ≈ (total_sum / 2) - s_left
                target = half_total - s_left

                # Binary search for 'target' in the sorted right sums to find lower bound
                import bisect
                idx = bisect.bisect_left(right_k_sums, target)
                # now we need to check for both idx between which out target is sandwiched to get to know the closest value to target

                # Check the value at the insertion point (if valid)
                if idx<len(right_k_sums):
                    s_right = right_k_sums[idx]
                    current_subset_sum = s_left + s_right
                    current_diff = abs(total_sum-2*current_subset_sum)
                    min_diff = min(min_diff,current_diff)
                # Check the value immediately before the insertion point (if valid)
                if idx > 0:
                    s_right = right_k_sums[idx - 1]
                    current_subset_sum = s_left + s_right
                    current_diff = abs(total_sum - 2 * current_subset_sum)
                    min_diff = min(min_diff, current_diff)
                # Early exit optimization: if we found a 0 difference, we can't do better
                if min_diff == 0:
                    return 0
        return min_diff
```

Count subsets with sum K

```python
class Solution:
    def perfectSum(self, arr, K):
        dp = dict()
        def helper(idx,target):
            if (idx,target) in dp:
                return dp[(idx,target)]
            if target==0:
                return 1
            
            if idx < 0 or target < 0:
                return 0
          
            
            # pick
            pick = 0
            if arr[idx]<=target:
                pick = helper(idx-1,target-arr[idx])
            # not pick
            not_pick = helper(idx-1,target)

            dp[(idx,target)] = pick+not_pick
            return pick+not_pick
        return helper(len(arr)-1,K)
    
Solution().perfectSum( arr = [1, 2, 3, 4, 5], K = 5)
```
