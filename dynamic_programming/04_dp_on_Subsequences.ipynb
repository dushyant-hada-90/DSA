```python
class Solution:
    def func(self,arr,k):
        n = len(arr)

        dp = [[float("-inf")]*(k+1) for _ in range(n)]
        def helper(idx,target):
                if dp[idx][target]!=float("-inf"):
                     return dp[idx][target]
                if target==0:
                    return True
                if idx<0:
                     return False
                # either take it
                # only if it is smaller than target 
                take = False  #you need to define it before 
                if arr[idx]<=target:
                    take =  helper(idx-1,target-arr[idx])
                # or dont take it
                not_take = helper(idx-1,target)
                dp[idx][target] = take or not_take 
                return dp[idx][target]

        return helper(n-1,k)
# o(n*k) since there are n elements and max value of target is k so total n*k states are possible # time comp would have been 2^n without dp
# o(n*k) for dp 
# o(n) stack space as at each level we handle a index 
Solution().func([2,3,1,1],4)
```

Partition Set Into 2 Subsets With Min Absolute Sum Diff 

```python
import itertools
class Solution:
    def minimumDifference(self, nums):
        n = len(nums) // 2
        total_sum = sum(nums)

        # split into two halves
        left_part = nums[:n]
        right_part = nums[n:]

        # Helper to generate all subset sums grouped by the number of elements (k)
        # Returns a dictionary: { count_of_elements: [list_of_sums] }
        def get_sums(arr):
            sums_by_count = {}
            # iterate through all lengths from 0 to N
            for k in range(len(arr)+1):
                sums_by_count[k] = []
                for seq in itertools.combinations(arr,k):#generates all unique subsequences of k items from arr 
                    sums_by_count[k].append(sum(seq))
            return sums_by_count
        
        left_sums = get_sums(left_part)
        right_sums = get_sums(right_part)

        # sort right_sums for binary search
        for k in right_sums:
            right_sums[k].sort()
        
        min_diff = float("inf")
        half_total = total_sum/2 #we want out subset to be as close as possible to half_total as possible

        for k in range(n+1):
            left_k_sums = left_sums[k]
            # If we took k from left, we need n-k from right
            right_target_count = n-k
            right_k_sums = right_sums[right_target_count]

            # for each sum in the left group, find the best match in the right group
            for s_left in left_k_sums:
                # We want: s_left + s_right ≈ total_sum / 2
                # So: s_right ≈ (total_sum / 2) - s_left
                target = half_total - s_left

                # Binary search for 'target' in the sorted right sums to find lower bound
                import bisect
                idx = bisect.bisect_left(right_k_sums, target)
                # now we need to check for both idx between which out target is sandwiched to get to know the closest value to target

                # Check the value at the insertion point (if valid)
                if idx<len(right_k_sums):
                    s_right = right_k_sums[idx]
                    current_subset_sum = s_left + s_right
                    current_diff = abs(total_sum-2*current_subset_sum)
                    min_diff = min(min_diff,current_diff)
                # Check the value immediately before the insertion point (if valid)
                if idx > 0:
                    s_right = right_k_sums[idx - 1]
                    current_subset_sum = s_left + s_right
                    current_diff = abs(total_sum - 2 * current_subset_sum)
                    min_diff = min(min_diff, current_diff)
                # Early exit optimization: if we found a 0 difference, we can't do better
                if min_diff == 0:
                    return 0
        return min_diff
```

```python
class Solution:
    def perfectSum(self, arr, K):
        dp = dict()
        def helper(idx,target):
            if (idx,target) in dp:
                return dp[(idx,target)]
            
            
            if idx < 0:
                return 1 if target==0 else 0
          
            
            # pick
            pick = 0
            if arr[idx]<=target:
                pick = helper(idx-1,target-arr[idx])
            # not pick
            not_pick = helper(idx-1,target)

            dp[(idx,target)] = pick+not_pick
            return pick+not_pick
        return helper(len(arr)-1,K)
    
Solution().perfectSum( arr = [0,1, 2, 3, 4, 5], K = 5)
```

```python
class Solution:
    def func(self,wt,val,capacity):
        n = len(val)
        dp = {}
        def helper(i,cap):
            if i<0 or cap<=0:
                return 0
            
            if (i, cap) in dp:
                return dp[(i, cap)]
            
            # either pick only if u have the capacity to pick
            pick = float("-inf")
            if cap>=wt[i]:
                pick = helper(i-1,cap-wt[i]) + val[i]
            # or not pick
            not_pick = helper(i-1,cap)
            dp[(i,cap)] = max(pick,not_pick)
            return dp[(i,cap)]


        return helper(n-1,capacity)
Solution().func([3,2,5],[30,40,60],6)
```

```python
class Solution:
    def func(self,wt,val,capacity):
        n = len(val)
        dp = [[0]*(capacity+1) for _ in range(n+1) ]

        # Row 0: This represents having zero items to pick from. No matter how big your backpack is, if there are no items, your value is 0.
        # Column 0: This represents having a capacity of 0. Even if you have a diamond in front of you, if your bag has no space, your value is 0.


        # Imagine you are filling the cell for Row i (the current item) and Capacity w:
        # Option A: Leave it. You look at the cell directly above you dp[i-1][w]. This is the best you did with the previous items at this same capacity.
        # Option B: Take it. You can only do this if the item's weight wt[i-1] fits in your current capacity w. If it fits, you take its value val[i-1] and add it to the best you could do with the remaining space: dp[i-1][w - wt[i-1]].

        for i in range(1,n+1):
            curr_wt = wt[i-1]
            curr_val = val[i-1]
            for cap in range(1,capacity+1):
                # pick
                pick = 0
                if curr_wt<=cap:
                    pick = curr_val + dp[i-1][cap - curr_wt]

                # not pick
                not_pick = dp[i-1][cap]

                dp[i][cap] = max(pick,not_pick)
        return dp[n][capacity]
Solution().func([3,2,5],[30,40,60],6)
```

```python
class Solution:
    def func(self, wt, val, capacity):
        n = len(val)

        dp = [0 for _ in range(capacity + 1)]

        # Build the table bottom-up
        for i in range(1, n + 1):
            curr_wt = wt[i-1]
            curr_val = val[i-1]
            for cap in range( capacity,curr_wt - 1,-1): #because you can't pick an item if the capacity is smaller than its weight anyway
                
                # Option 1: Not picking the current item
                not_pick = dp[cap]
                
                # Option 2: Picking the current item (if weight allows)
                pick = 0
                if curr_wt <= cap:
                    pick = curr_val + dp[cap - curr_wt]
                
                # Store the maximum of both choices
                dp[cap] = max(pick, not_pick)

        return dp[capacity]

# Testing with your values
sol = Solution()
result = sol.func([3, 2, 5], [30, 40, 60], 6)
print(f"Maximum Value: {result}")
```

```python
class Solution:
    def coinChange(self,coins,target):
        n = len(coins)
        dp = [[-1] * (target + 1) for _ in range(n)]
        def helper(i,rem):
            if rem==0:
                return 0 #you cant pick a coin anywhere further if rem==0 , basically you have created a target
            if i<0 or rem<0:
                return float("inf") #to signify that amount cant be created by this
            if dp[i][rem]!=-1:
                # print("early return")
                return dp[i][rem]
            # pick again the same coin
            # pick only if it does not exceed target
            pick = float("inf")
            if coins[i]<=rem:
                pick = helper(i,rem-coins[i])+1
            # or not pick this one and move to next
            not_pick = helper(i-1,rem)
            dp[i][rem] = min(pick,not_pick)
            return dp[i][rem]
        ans = helper(n-1,target)
        return ans if ans!=float("inf") else -1
# exp time complexity without dp 

# O(target*n) 
# o(target*n) and stack space o(n)
Solution().coinChange([411,412,413,414,415,416,417,418,419,420,421,422],9864)
```

Target Sum

```python
class Solution:
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        # so basically i can multiply a number either py +1 or -1 and the final sum has to reach targget, so i will try out all possibilities
        # there wont be pick not pick case as we have to pick only.

        dp = {}
        n = len(nums)
        def helper(i,tar):
            if (i,tar) in dp:
                return dp[(i,tar)]
            # Base Case: When all numbers are exhausted
            if i < 0:
                # If we reached the target, this path is valid (return 1)
                return 1 if tar == 0 else 0
            # multiply by +1 and pick
            pos = helper(i-1,tar-nums[i]*1)
            # multiply by -1 and pick
            neg = helper(i-1,tar-nums[i]*-1)

            dp[(i,tar)] = pos+neg
            return dp[(i,tar)]
        return helper(n-1,target)
Solution().findTargetSumWays(nums = [1,1,1,1,1], target = 3)
```

```python
# But we can solve this better by finding a subset such that it sums up to:

# sum(P) - sum(N) = target
# sum(P) + sum(N) = total_sum
# 
# Adding these two equations:
# 2*sum(P) = target + total_sum
# sum(P) = {target + total_sum}/{2}
# now find a subset with that sum

class Solution:
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        n = len(nums)
        total_sum = sum(nums)

        if abs(target)>total_sum or (total_sum+target)%2 != 0:
            return 0
        k = (total_sum+target)//2
        
        
        dp = [[-1]*(k+1) for _ in range(n)]
        def helper(i,tar):
            if dp[i][tar]!=-1:
                return dp[i][tar]
            
            if i<0:
                return 1 if tar==0 else 0
                
            # pick
            pick = 0
            if nums[i]<=tar: #ensuring that for next call tar is not negative (Note: still tar can be zero which is handled in base condition)
                pick = helper(i-1,tar-nums[i])
            # not_pick
            not_pick = helper(i-1,tar)

            dp[i][tar] = pick+not_pick
            return dp[i][tar]
        return helper(n-1,k)
    
Solution().findTargetSumWays(nums = [1,1,1,1,1], target = 3)
```

```python
from collections import Counter
# will visit again
class Solution:
    def findTargetSumWays(self, nums, t):
        n = len(nums)
        left_part = nums[:n//2]
        right_part = nums[n//2:]

        def get_sums_dict(arr):
            # Start with 1 way to get a sum of 0
            dp = Counter({0: 1})
            for x in arr:
                new_dp = Counter()
                for s, count in dp.items():
                    # Every existing sum 's' now branches into two:
                    new_dp[s + x] += count # The +x branch
                    new_dp[s - x] += count # The -x branch
                dp = new_dp
            return dp
        
        left_sums = get_sums_dict(left_part)
        right_sums = get_sums_dict(right_part)
        
        count = 0
        # For every sum we found in the left half...
        for l_sum, l_count in left_sums.items():
            needed = t - l_sum
            # ...check if the "missing piece" exists in the right half
            if needed in right_sums:
                # Multiply the ways: (Ways to get L) * (Ways to get R)
                count += l_count * right_sums[needed]
               
        return count
Solution().findTargetSumWays(nums = [1,1,1,1,1], t = 4)
```

```python
class Solution:
    def unboundedKnapsack(self, wt, val, W):
        n = len(val)
        dp = {}
        def helper(i,cap):
            if i<0 or cap<=0:
                return 0
            
            if (i, cap) in dp:
                return dp[(i, cap)]
            
            # either pick only if u have the W to pick
            pick = float("-inf")
            if cap>=wt[i]:
                pick = helper(i,cap-wt[i]) + val[i]
            # or not pick
            not_pick = helper(i-1,cap)

            dp[(i,cap)] = max(pick,not_pick)
            return dp[(i,cap)]


        return helper(n-1,W)
# o(n*w)
# o(n*w)
Solution().unboundedKnapsack(val = [10, 40, 50, 70], wt = [1, 3, 4, 5], W = 8)
```

```python
class Solution:
    def unboundedKnapsack(self, wt, val,  W):
        n = len(val)
        dp = [[0]*(W+1) for _ in range(n+1) ]

        # Row 0: This represents having zero items to pick from. No matter how big your backpack is, if there are no items, your value is 0.
        # Column 0: This represents having a W of 0. Even if you have a diamond in front of you, if your bag has no space, your value is 0.


        # Imagine you are filling the cell for Row i (the current item) and W w:
        # Option A: Leave it. You look at the cell directly above you dp[i-1][w]. This is the best you did with the previous items at this same W.
        # Option B: Take it. You can only do this if the item's weight wt[i-1] fits in your current W w. If it fits, you take its value val[i-1] and add it to the best you could do with the remaining space: dp[i-1][w - wt[i-1]].

        for i in range(1,n+1):
            curr_wt = wt[i-1]
            curr_val = val[i-1]
            for cap in range(1,W+1):
                # pick
                pick = 0
                if curr_wt<=cap:
                    pick = curr_val + dp[i][cap - curr_wt]

                # not pick
                not_pick = dp[i-1][cap]

                dp[i][cap] = max(pick,not_pick)
        return dp[n][W]
Solution().unboundedKnapsack(val = [10, 40, 50, 70], wt = [1, 3, 4, 5], W = 8)
```

```python
class Solution:
    def unboundedKnapsack(self, wt, val,  W):
        n = len(val)

        dp = [0 for _ in range(W + 1)]

        # Build the table bottom-up
        for i in range(1, n + 1):
            curr_wt = wt[i-1]
            curr_val = val[i-1]
            for cap in range(W+1): 
                
                # Option 1: Not picking the current item
                not_pick = dp[cap]
                
                # Option 2: Picking the current item agin (if weight allows)
                pick = 0
                if curr_wt <= cap:
                    pick = curr_val + dp[cap - curr_wt]
                
                # Store the maximum of both choices
                dp[cap] = max(pick, not_pick)

        return dp[W]


sol = Solution()
Solution().unboundedKnapsack(val = [10, 40, 50, 70], wt = [1, 3, 4, 5], W = 8)
```

Rod cutting problem

```python
class Solution:
    def RodCutting(self, price, n):
        dp = [0]*(n+1)
        # dp[i] stores the maximum value obtainable for a rod of length i
        dp[0] = 0 #if i do cut a rope of len==0 then i will earn 0 rupees 

        # Iterate through each possible cut length 'i' (1 to n)
        # When i=1, you are asking: "How much value can I get if I'm allowed to use pieces of length 1?"
        # When i=2, you are asking: "How much better can I do if I now also allow pieces of length 2?"
        for i in range(1, n + 1):
            curr_price = price[i-1]
            # Update the dp table for all rod lengths 'j' that can accommodate length 'i'
            for j in range(i, n + 1): #j represents length of rod which at the end becomes n to give us our answer
                # We either keep the current max at dp[j] (not_cut)
                # or we take the piece of length 'i' and add it to the max value 
                # of the remaining length 'j-i' (cut)
                dp[j] = max(dp[j], curr_price + dp[j - i])
        return dp[n]
Solution().RodCutting(price = [1, 6, 8, 9, 10, 19, 7, 20], n = 8)
                    
```
