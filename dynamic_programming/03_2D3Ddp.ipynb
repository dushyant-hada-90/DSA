```python
'''
suppose we have to plan for 10 days such that we can earn max merit points  then the task to be done on 8th day will depend on the task done on 9th day and similarly taks done on 7th day will depend on 8th day 

if i find max merit points till 8 th day then we can find max merit points till 9th day by points_on_9th = max(arr[9][i] + helper(8,i)) run i for all possibilitties of tasks on 9th day

sample helper call-> helper(calaculate max points possible for this day, given that this task was done on n+1 th day)
we are starting our traversal from nth day to 0th day 

all three tasks can be denoted by 0,1,2
-1 means that not ask was done
'''
# memoization
class Solution:
    def ninja(self,arr):
        dp = {}
        def helper(day,last):

            if (day,last) in dp:
                return dp[(day,last)]
            print(day,last)
            
            if day==0:
                maxi = float("-inf")
                for i in range(3):
                    if i!=last:
                        maxi = max(maxi,arr[day][i])
                dp[(day,last)] = maxi
                return maxi
            
            maxi = float("-inf")
            for i in range(3):
                if i!= last:
                    points = arr[day][i] + helper(day-1,i) #points represents max possible earned points till n-th day
                    # we are exploring all possibilities by trying both tasks on nth day and seeing which way we can earn more points overall
                    maxi = max(maxi,points)
            dp[(day,last)] = maxi
            return maxi

        n = len(arr)
        return helper(n-1,-1)
# o(n*4*3) ->4 is in worst case for possibilities of last(generally its 3) and n is the umber of days so there can be 4n recursive calls in worst case after implemnting dp  and each funcition has a for loop inside that runs 3 times 
# o(n) stack space , o(n*4) for dp
Solution().ninja([[70, 40, 10], [180, 20, 5], [200, 60, 30]])
```

```python
# tabulation
# this time we start from 0th index and go to n-1 th day
class Solution:
    def ninja(self, arr):
        n = len(arr)

        # Create a DP table:
        # dp[day][last] â†’ max points till 'day' if last task done previously was 'last'
        # 4 because last can be 0,1,2 (actual tasks) or 3 (no restriction)
        dp = [[0]*4 for _ in range(n)]

        # -----------------------------
        # BASE CASE (Day 0)
        # -----------------------------
        # On the first day, there is no "previous" task.
        # For every possible 'last' value, choose the max points among tasks
        # that are NOT equal to 'last'.
        for last in range(4):
            dp[0][last] = max(
                arr[0][task] for task in range(3) if task != last
            )

        # ðŸ§© Visualization after this base case:
        # dp[0][0] â†’ max points on day 0 if task 0 was NOT done previously
        # dp[0][1] â†’ max points if task 1 was NOT done previously
        # dp[0][2] â†’ max points if task 2 was NOT done previously
        # dp[0][3] â†’ max points if thereâ€™s no restriction at all
        #
        # In simple words, dp[0][3] will just be max(arr[0]) because nothing was done before day 0.

        # -----------------------------
        # FILL DP TABLE (from day 1 â†’ n-1)
        # -----------------------------
        for day in range(1, n):
            print(f"\nDay {day}:")
            print("Previous day's DP:", dp[day-1])

            # For each possible 'last' task (0,1,2,3)
            for last in range(4):
                # Try doing each task (0,1,2) today, except the one done yesterday (last)
                best = float('-inf')

                for task in range(3):
                    if task != last:
                        # Today's points + best possible till yesterday
                        points_today = arr[day][task] + dp[day-1][task]

                        # Keep track of best among all choices
                        best = max(best, points_today)
                        print(
                            f"  If yesterday={last}, today={task}: "
                            f"points_today={arr[day][task]}, "
                            f"total={points_today}"
                        )

                dp[day][last] = best
                print(f"Best for last={last}: {dp[day][last]}")

        print("\nFinal DP Table:")
        for d in range(n):
            print(f"Day {d}: {dp[d]}")

        # -----------------------------
        # FINAL ANSWER
        # -----------------------------
        # dp[n-1][3] means â†’ last day, no restriction on what task we do.
        # It already stores the max possible points across all valid combinations.
        return dp[n-1][3]

Solution().ninja([[70, 40, 10], [180, 20, 5], [200, 60, 30]])
```
