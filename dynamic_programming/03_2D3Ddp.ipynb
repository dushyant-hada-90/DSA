Ninja's Training (DP 7)

```python
# memoization
class Solution:
    def ninja(self,arr):
        n = len(arr)
        dp = {}
        def helper(day,last):
            if (day,last) in dp.keys():
                return dp[(day,last)]

            if day == 0:
                max_points = float("-inf")
                for i in range(3):
                    if i!= last:
                        max_points = max(max_points,arr[0][i])
                return max_points
            
            max_points = float("-inf")
            for i in range(3):
                if i!= last:
                    points = arr[day][i] + helper(day-1,i)
                    max_points = max(max_points,points)
            dp[(day,last)] = max_points
            return max_points

        return helper(n-1,3)
        
# o(n*4*3) ->4 is in worst case for possibilities of last(generally its 3) and n is the umber of days so there can be 4n recursive calls in worst case after implemnting dp  and each funcition has a for loop inside that runs 3 times 
# o(n) stack space , o(n*4) for dp
sol = Solution()
ans = sol.ninja([[92, 14, 55], [3, 78, 21], [60, 4, 99], [17, 45, 82], [33, 91, 12], [70, 2, 68], [44, 25, 59], [88, 10, 30], [5, 52, 77], [29, 95, 41], [1, 63, 85], [100, 38, 7], [16, 49, 90], [22, 73, 5], [51, 8, 35], [97, 66, 19], [13, 31, 74], [62, 80, 48], [36, 94, 28], [57, 18, 93]])
print(ans)
```

Grid Unique Paths : DP on Grids (DP8)

```python
# memoization
class Solution:
    def uniquePaths(self,m,n):
        print(m,n)
        
        dp = {(0,0):1}
        def helper(i,j):
            if (i,j) in dp.keys():
                return dp[(i,j)]
            
            if i<0 or j<0 :
                return 0
            
            top = helper(i-1,j)
            left = helper(i,j-1)
            dp[(i,j)] = top+left
            
            return top+left
        return helper(m-1,n-1)
Solution().uniquePaths([[1,0]])
```

Minimum Path Sum

```python
# graph approach
import heapq
class Solution:
    def minPathSum(self, grid):
        n = len(grid)
        m = len(grid[0])
        heap = [(0,0,0)]
        delta = ((1,0),(0,1))
        dist = [[float("inf")]*m for _ in range(n)]
        dist[0][0] = grid[0][0]

        while heap:
            sum_carried,i,j = heapq.heappop(heap)
            for di,dj in delta:
                row,col = i+di,j+dj
                if 0<=row<n and 0<=col<m:
                    weight = grid[row][col]
                    if  sum_carried + weight < dist[row][col]:
                        dist[row][col] = dist[i][j] + weight
                        heapq.heappush(heap,(dist[i][j] + weight,row,col))
        return dist[-1][-1]

#o(n*m)(log(n*m))
# o(n*m) for dist and o(m*n) in worst case for heap   
Solution().minPathSum([[1,2,3],[4,5,6]])
```

```python
# top down
class Solution:
    def func(self,triangle):
        n = len(triangle)
        dp = {}
        def helper(i,j):
            if (i,j) in dp:
                return dp[(i,j)]
            if j>=len(triangle[i]): # not possble to got hat right
                return float("inf")
            if i==n-1:
                return triangle[i][j]
            # either go bottom
            bottom = helper(i+1,j)
            # or go right-bottom
            bottom_right = helper(i+1,j+1)


            # store before returning
            dp[(i,j)] = triangle[i][j] + min(bottom_right,bottom)
            return triangle[i][j] + min(bottom_right,bottom)

        return helper(0,0)
# o(2^N) if dp not implemented but after implementing its o(N) each time only new unique node is processed , here N is total items in triangle, o(N) stack space is also used
# o(N) for dp
Solution().func(triangle = [[2],[3,4],[6,5,7],[4,1,8,3]])
```

```python
# top down
class Solution:
    def func(self,matrix):
        n = len(matrix)
        dp = {}

        def helper(i,j):
            if (i,j) in dp:
                return dp[(i,j)]
            if j>=n or j<0: # not possble to got hat right
                return float("inf")
            if i==n-1:
                return matrix[i][j]
            # either go bottom
            bottom = helper(i+1,j)
            # or go right-bottom
            bottom_right = helper(i+1,j+1)
            # or go left bottom
            bottom_left = helper(i+1,j-1)


            # store before returning
            dp[(i,j)] = matrix[i][j] + min(bottom_right,bottom,bottom_left)
            return matrix[i][j] + min(bottom_right,bottom,bottom_left)
        ans = float("inf")
        for col in range(n):
            ans =  min(ans,helper(0,col))
        return ans
# o(3^n^2) if dp not implemented but after implementing its o(n^2) each time only new unique node is processed , here n is len of matrix[0], o(n^2) stack space is also used
# o(n^2) for dp 
Solution().func(matrix = [[2,1,3],[6,5,4],[7,8,9]])
```

```python
# bottom up
class Solution:
    def func(self,matrix):
        n = len(matrix)
        dp = {}

        for index,value in enumerate(matrix[-1]):
            dp[(n-1,index)] = value

        for i in range(n-2,-1,-1):
            for j in range(n-1,-1,-1):
                # either go bottom
                bottom = dp.get((i+1,j),float("inf"))
                # or go right bottom_right
                bottom_right = dp.get((i+1,j+1),float("inf"))
                # or go bottom left
                bottom_left = dp.get((i+1,j-1),float("inf"))

                dp[(i,j)] = matrix[i][j] + min(bottom,bottom_right,bottom_left)
        
        return min(dp[0, i] for i in range(n))
# o(n^2) where n is len of matrix
# o(n^2) for dp
Solution().func(matrix = [[2,1,3],[6,5,4],[7,8,9]])
```

```python
# space optimized
class Solution:
    def func(self, matrix):
        dp = matrix[-1][:]  
        n = len(matrix)

       
        for i in range(len(matrix) - 2, -1, -1):
            prev = dp[:]
            for j in range(n-1,-1,-1): #it does not really matter the direction in which j runs
                dp[j] = matrix[i][j] + min(prev[j],
                                           prev[j-1] if j>0 else float("inf"),prev[j+1] if j<n-1 else float("inf"))

        return min(dp)

# o(n^2) where n^2 is total items in matrix
# o(m) where m is len of last row
Solution().func(matrix = [[2,1,3],[6,5,4],[7,8,9]])
```

```python
# top down
class Solution:
    def func(self,grid):
        n = len(grid) #square grid
        dp = [[[-1]*n for _ in range(n)] for _ in range(n)]

        def helper(i,j1,j2):
            if j1<0 or j1>=n or j2<0 or j2>=n:
                return float("-inf")
            
            if dp[i][j1][j2]!=-1:
                return dp[i][j1][j2]
            
            if i==n-1:
                if j1==j2:
                    return grid[i][j1]
                else:
                    return grid[i][j1]+grid[i][j2]
                
            best = float("-inf")
            for dj1 in [-1, 0, 1]:
                for dj2 in [-1, 0, 1]:
                    best = max(best,helper(i+1,j1+dj1,j2+dj2))
            
            curr = grid[i][j1]
            if j1!=j2:
                curr += grid[i][j2]
            dp[i][j1][j2] = curr + best

            return dp[i][j1][j2]

        return helper(0,0,n-1)
# o(9*n^3) because there can be diff values for i,j1,j2 as input to function each of them ranging from 0  to n and for each value the loop runs 9 times 
#o(n^3) for dp and o(n) for stack space because at each depth we are processing a row and there are n rows so at max there will be n functions in stack
Solution().func(grid = [[2, 3, 1, 2],[3, 4, 2, 2],[5, 6, 3, 5]])
```
