Longest Increasing Subsequence

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [[-1 for _ in range(n+1)] for _ in range(n)]#prev starts from -1

        def helper(i,prev):
           
            if i==n:
                return 0
            
            # Jab bhi DP table check karenge ya store karenge, prev ki jagah prev+1 use karenge.
            if dp[i][prev+1]!=-1: return dp[i][prev+1]

            # take if you can
            take = float("-inf")
            if prev==-1 or nums[i]>nums[prev]:
                take = helper(i+1,i) + 1
            # not take
            not_take = helper(i+1,prev)

            dp[i][prev+1] = max(take,not_take)

            return dp[i][prev+1]
         
        return helper(0,-1)
# o(n*n)
# o(n*n) dp and o(n) stack space
Solution().lengthOfLIS([10,9,2,5,3,7,101,18])
```

```python
# "Classic 1D DP" ka logic simple hai: Har element ko ek potential "Hero" (Ending point) maano.
# jab len nikalni ho toh copressed 2d dp wali approach kaam karegi aur ye bhi but jab specific questions mein path reconstruct karna ho tab yahi approach easy hai
# "Agar meri sequence index i par khatam hoti hai, toh piche wale kis element ke saath judne par mujhe sabse lambi chain milegi?"
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        if not nums: return 0
        n = len(nums)

        # dp[i] = Longest Increasing Subsequence ending EXACTLY at index i
        # Base Case: Har element khud mein 1 length ki sequence hai
        dp = [1] * n

        # Forward Loop: i ko end point maante hue aage badho
        for i in range(n):
            # i ke piche wale saare elements (j) ko check karo
            for j in range(i):
                if nums[i]>nums[j]:# Agar nums[i] piche wale nums[j] se bada hai
                    dp[i] = max(dp[i],1+dp[j])
        return max(dp)

Solution().lengthOfLIS([10,9,2,5,3,7,101,18])
```

```python
# pro nlogn solution
import bisect

class Solution:
    def lengthOfLIS(self,nums:list[int])->int:
        tails = []

        for x in nums:
            # binary search se dhundo ki x kis pile mein fit hoga
            # bisect_left se lower bound nikal lo
            idx = bisect.bisect_left(tails,x)

            if idx == len(tails): #implying ki existing piles mein fit nahi ho paa rha toh hum new pile  banaenge
                tails.append(x)
            else:
                # agar x kisi pile mein fit ho gaya toh hum uss pile ke top item ko x se replace kar denge
                tails[idx] = x
        return len(tails)
    
# Hum array par ek baar iterate karte hain (n).
# Har element ke liye bisect_left (Binary Search) karte hain (log n).
# Total: n times log n => O(n log n).
Solution().lengthOfLIS([10,9,2,5,3,7,101,18])

# "Best" ka matlab hai khud ko itna chota rakhna ki nums ke aage wale elements ke liye jagah ban sake.
# tails[k] ka matlab hai: "Sabse chota element jo length k+1 ki LIS khatam kar sakta hai."
```

Printing Longest Increasing Subsequence

```python
import bisect
class Solution:
    def printLIS(self,nums:list[int])->list[int]:
        if not nums : return []
        n = len(nums)
        tails_indices = []
        parent = [-1]*n

        for i in range(n):
            # we cant use bisect because now tails contains index of items of nums not the values of items of nums
            current_val = nums[i]

            # find lower bound manually
            low,high = 0,len(tails_indices)-1
            while low<=high:
                mid = (low+high)//2
                if nums[tails_indices[mid]]<current_val:
                    low = mid+1
                else:
                    high = mid-1
            idx = low

            # Agar humne length > 1 ki sequence banayi hai (idx > 0)
            # toh iska parent wo hoga jo isse pehle waale pile ke top par tha
            if idx > 0:
                parent[i] = tails_indices[idx-1]
            
            if idx == len(tails_indices):
                tails_indices.append(i)
            else:
                tails_indices[idx] = i
        # backtrack to construct lis
        lis = []
        curr = tails_indices[-1]
        while curr!=-1:
            lis.append(nums[curr])
            curr = parent[curr]
        return lis[::-1]
Solution().printLIS([10,9,2,5,3,7,101,18])

# Jab bhi koi naya element (nums[i]) kisi pile mein jaata hai, hum use bolte hain: "Tera pichla saathi (parent) woh hai jo abhi tere se left waale pile ke top par baitha hai."

# Kyunki left waala pile hamesha usse chhota element hi hold karega (sorted property), humein hamesha ek valid increasing connection mil jata hai.

# Ek Mazedaar Baat: Is logic mein bhale hi tails mein elements badalte rahein, lekin parent array purani memory ko delete nahi hone deti. Woh ek Linked List ki tarah piche ka rasta yaad rakhti hai.
```

Largest Divisible Subset

```python
class Solution:
    def largestDivisibleSubset(self, nums: List[int]) -> List[int]:
        n = len(nums)

        def helper(i,prev):
            # if i exhaust then return 0
            if i==n:
                return 0
            # take
            take = float("-inf")
            if prev==-1 or  nums[i]%nums[prev]==0:
                take = helper(i+1,i)+1
            elif nums[prev]%nums[i]==0:
                take = helper(i+1,prev)+1
            # not take
            not_take = helper(i+1,prev)


            return max(take,not_take)
        return helper(0,-1)
# agar isme parent wala concept lagaunga toh har call ke end mein new list return karni padegi which will be memory heavy 
Solution().largestDivisibleSubset(nums = [1,2,4,8])
```

```python
class Solution:
    def largestDivisibleSubset(self, nums: List[int]) -> List[int]:
        n = len(nums)
        if n == 0: return []
        
        # Step 1: Sorting bohot zaruri hai taaki humein sirf nums[i] % nums[j] check karna pade
        # Agar a < b aur b % a == 0, toh aage aane wala c agar b ko divide karega toh a ko automatically karega
        nums = sorted(nums)
        
        # dp[i] = "Max length of divisible subset jo index 'i' par khatam ho rahi hai"
        # Initial value 1 kyunki har element akela 1 length ka subset toh hai hi
        dp = [1 for _ in range(n)]
        
        # parent[i] = "Index 'i' se pehle wala element kaunsa tha" (path yaad rakhne ke liye)
        parent = [-1] * n

        max_len = 0 # Poori array mein sabse lambi subset ki length track karne ke liye
        curr = 0    # Woh index jahan sabse lambi subset khatam hui

        # Step 2: Har element 'i' ko ending point maan kar check karo
        for i in range(n):
            # 'i' se pehle wale saare elements 'j' ko scan karo
            for j in range(i):
                # Agar nums[i] pichle nums[j] se divide hota hai
                if nums[i] % nums[j] == 0:
                    # Aur agar 'j' ke aage 'i' ko jodne se nayi lambi sequence mil rahi hai
                    if dp[j] + 1 > dp[i]:
                        dp[i] = dp[j] + 1
                        parent[i] = j # Yaad rakho ki 'i' ka pichla saathi 'j' tha
            
            # Har iteration ke baad global max length update karo
            if dp[i] > max_len:
                max_len = dp[i]
                curr = i # Yeh hamara "End Door" hai jahan se backtracking shuru hogi

        # Step 3: Backtracking (Parent pointers follow karke rasta nikaalo)
        ans = []
        while curr != -1: # Jab tak -1 na mil jaye (starting point)
            ans.append(nums[curr])
            curr = parent[curr] # Pichle element par jao
            
        # Kyunki humne piche se (bade se chota) add kiya hai, isliye reverse kardo
        return ans[::-1]

# Example Input: [1, 2, 4, 9, 8, 9]
# Output: [1, 2, 4, 8]
```

Longest String Chain

```python
class Solution:
    def longestStrChain(self, words: List[str]) -> int:
        n = len(words)
        words.sort(key=len)

        def compare(word1,word2):
            i,j = 0,0
            if len(word2)-1!=len(word1):return False

            while i<len(word1) and j < len(word2):
                if word1[i]==word2[j]:
                    i+=1
                    j+=1
                else: j+=1
            if (i==len(word1) and j==len(word2)) or j==len(word2)-1: return True
            return False

        dp = [[-1 for _ in range(n+1)] for _ in range(n)]
        def helper(i,prev_i):
            if i==n:
                return 0 # words exhausted
            
            if dp[i][prev_i]!=-1:return dp[i][prev_i]
            
            
            # take
            take = float("-inf")
            # c = compare(words[prev_i],words[i])
            # print(c,words[prev_i],words[i])
            if (prev_i == -1) or compare(words[prev_i],words[i]):
                take = helper(i+1,i) + 1
            # not take
            not_take = helper(i+1,prev_i)

            return max(take,not_take)
        return helper(0,-1)
Solution().longestStrChain( words = ["xbc","pcxbcf","xb","cxbc","pcxbc"])
```

```python
# 1d dp classic solution
class Solution:
    def longestStrChain(self, words: List[str]) -> int:
        if not words: return 0
        # Now we ensure shorter words come before longer words.
        words.sort(key=len)
        n = len(words)

        def compare(word1, word2):
            if len(word2) != len(word1) + 1: return False
            i, j = 0, 0
            while i < len(word1) and j < len(word2):
                if word1[i] == word2[j]:
                    i += 1
                j += 1
            return i == len(word1)

        # dp[i] = Longest Increasing Subsequence ending EXACTLY at index i
        # Base Case: Har element khud mein 1 length ki sequence hai
        dp = [1] * n

        # Forward Loop: i ko end point maante hue aage badho
        for i in range(n):
            # i ke piche wale saare elements (j) ko check karo
            for j in range(i):
                if compare(words[j], words[i]):# Agar words[i] piche wale words[j] se bada hai
                    dp[i] = max(dp[i],1+dp[j])
        return max(dp)
Solution().longestStrChain( words = ["xbc","pcxbcf","xb","cxbc","pcxbc"])
```

```python
# better approach
# Because words are very short ($L=16$), generating all possible predecessors is much faster than checking every other word in a long list ($n=1000$).

class Solution:
    def longestStrChain(self, words: List[str]) -> int:
        words.sort(key=len)

        # dp stores: { word : max_chain_length_ending_here }
        dp = {}
        max_overall = 0

        for word in words: #o(n)
            best_for_this_word = 1 #it can start a chain from itself

            # 2. Try removing every character one by one
            for i in range(len(word)): #o(l)
                # create predecessor by skipping char at index i
                predecessor = word[:i] + word[i+1:] #o(l)

                # If this smaller word has been seen before then we found a chain
                if predecessor in dp:
                    best_for_this_word = max(best_for_this_word,dp[predecessor]+1)
            dp[word] = best_for_this_word

            max_overall = max(max_overall,dp[word])

        return max_overall
# o(n*l^2) still better than o(n^2) because generallly l is very small than n , l is len of longest possible word in words
# sorting takes nlogn, outer loop runs n times , inner loop runs l times and string slicing also consumes l time at worst

# O(n*l) space complexity
# o(sorting takes o(n) space, dp takes o(n*l) space , string slicing takes o(l) because it is a temporary space and is overwritten at each iteration of innerlopp so space consumed can be at max the len of longest word

Solution().longestStrChain( words = ["xbc","pcxbcf","xb","cxbc","pcxbc"])
```

Longest Bitonic Subsequence

```python
class Solution:
    def func(self, nums: list[int]) -> int:
        if not nums: return 0
        
        # dp1[i] stores the length of the Longest Increasing Subsequence ending at i
        dp1 = self.get_lis(nums)
        
        # dp2[i] stores the length of the Longest Decreasing Subsequence starting at i
        # We find this by reversing the array, calculating LIS, then reversing the result back
        dp2 = self.get_lis(nums[::-1])[::-1]
        
        ans = 0
        # Combine the two sequences at each potential "peak" index i
        for i, j in zip(dp1, dp2):
            # i + j - 1: We subtract 1 because nums[i] is counted in both dp1 and dp2
            ans = max(ans, i + j - 1)
            
        return ans, dp1, dp2
    
    def get_lis(self, nums):
        n = len(nums)
        dp = [1] * n

        for i in range(n):
            for j in range(i):
                if nums[i] > nums[j]:
                    # Standard LIS update: can we extend the sequence ending at j?
                    dp[i] = max(dp[i], dp[j] + 1)
        return dp
    
# o(n^2) for finding lis dp
# o(n) for both dp arrays
# Example output for [1, 11, 2, 10, 4, 5, 2, 1]
# Peak at 5: [1, 2, 4, 5] (len 4) + [5, 2, 1] (len 3) - 1 = 6

Solution().func([9,8,1,7,6,5,4,3,2,1])
```

 Number of Longest Increasing Subsequence

```python
class Solution:
    def findNumberOfLIS(self, nums: List[int]) -> int:
        if not nums:return 0
        n = len(nums)
        dp = [(1,1)]*n

        for i in range(n):
            for j in range(i):
                if nums[i]>nums[j]:
                    # Case 1: Found a longer path
                    if dp[j][0]+1 > dp[i][0]:
                        dp[i] = (dp[j][0]+1,dp[j][1]) # Inherit the number of ways from j

                    # Case 2: Found another path of the SAME maximum length
                    elif dp[j][0]+1==dp[i][0]:
                        dp[i] = (dp[i][0],dp[i][1]+dp[j][1]) # You should add all the ways that index j reached that length.
        max_len = max(length for length,_ in dp)# max_len kya hai
        return sum(count for length,count in dp if length==max_len)#max_len kitni baar aa rhi hai
    
# o(n^2)
# o(n)
Solution().findNumberOfLIS(nums = [1,2,4,3,5,4,7,2])
```
