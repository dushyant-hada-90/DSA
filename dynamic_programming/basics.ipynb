Fibonacci number

```python
# memoization
# assuming that fibo sequence is 0 indexed
# sequence begins as 0, 1, 1, 2, 3
class Solution:
    def fibo(self,n):
        dp = {0:0,1:1}
        def helper(n):
            if n in dp:
                return dp[n]
            last = helper(n-1)
            second_last = helper(n-2)
            dp[n] = last+second_last
            return last+second_last
        return helper(n)
# o(n)
# o(n) stack space, o(n) dict space
Solution().fibo(2)
```

```python
# tabulation
# assuming that fibo sequence is 0 indexed
# sequence begins as 0, 1, 1, 2, 3
class Solution:
    def fibo(self,n):
        if n==0:
            return 0
        last,second_last = 1,0 #better to use these pointers instead of dict
        for i in range(2,n+1):
            updated_last = last+second_last
            second_last = last
            last =  updated_last
            # print(i,last)

        return last
# o(n)
# o(1)
Solution().fibo(10)
```
