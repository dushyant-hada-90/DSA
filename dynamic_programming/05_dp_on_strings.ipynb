Longest Common Substring

```python
class Solution:
    def longestCommonString(self, text1: str, text2: str) -> int:
        n = len(text1)
        m = len(text2)
        dp = [0 for _ in range(len(text2)+1)]
        ans = 0
        for i in range(1,n+1):
            prev_diagonal = 0
            for j in range(1,m+1):
                temp = dp[j]  
                if text1[i-1]==text2[j-1]:
                    dp[j] = 1 + prev_diagonal
                    ans = max(ans,dp[j])
                    # In the Substring problem, the "streak" can die at any time. The longest common part might happen in the middle of the strings and then get reset to zero by a later mismatch. Therefore, you must track the maximum value seen anywhere in the table.
                    
                else:
                    dp[j] = 0 # Streak Breaker
                 
                prev_diagonal = temp 
        return ans
# o(n*m)
# o(m)
Solution().longestCommonString(text1 = "abcde", text2 = "abce"  ) 
```
