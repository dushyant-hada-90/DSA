Matrix Chain Multiplication

```python
# in top down we went from a bigger problem (solving for array) to smaller problem (solving for one item)
class Solution:
    def func(self,arr):
        n = len(arr)
        dp = [[-1 for _ in range(n)] for _ in range(n)]
        def solve(i,j):
            if i==j:return 0
            if i+1==j:return arr[i-1]*arr[i]*arr[j]
            if dp[i][j]!=-1:return dp[i][j]

            ans = float("inf")
            for k in range(i,j):
                left = solve(i,k)
                right = solve(k+1,j)
                ans = min(ans,left+right+arr[i-1]*arr[k]*arr[j])
            dp[i][j] = ans

            return dp[i][j]
        return solve(1,n-1)
# o(n^3)
# o(n) stack space, o(n^2) dp space
Solution().func( [10, 15, 20, 25])
```

```python
# in bottom up we will go from small problem (one item ) to eventually dp[the]biggest problem(entire array)
class Solution:
    def func(self,arr):
        n = len(arr)
        dp = [[0 for _ in range(n)] for _ in range(n)]#0th row is buffer because items start from index 1 and go to n-1
        for i in range(n):
            dp[i][i] = 0 #no calculation needed
            
        for i in range(n-1,0,-1):
            for j in range(i+1,n): #reverso of what happened in recursion
                if i+1==j:
                    dp[i][j] =  arr[i-1]*arr[i]*arr[j]
                    continue

                ans = float("inf")
                for k in range(i,j):
                    left = dp[i][k]
                    right = dp[k+1][j]
                    ans = min(ans,left+right+arr[i-1]*arr[k]*arr[j])
                dp[i][j] = ans
        # tabulation loops ke last mein i will stop at 1 and j will stop at n-1
        return dp[1][n-1]
#  o(n^3)
# o(n) stack space, o(n^2) dp space
Solution().func( [10, 15, 20, 25])
```

Minimum Cost to Cut the Stick

```python
'''
Is problem mein dikkat ye hai ki har cut ka cost us stick ki current length hota hai, isliye hume cuts ka order smartly choose karna hota hai.
Maine socha: â€œAgar main kisi range (l, r) ke andar pehle kaunsa cut karun, to baaki ka problem apne-aap chhote parts mein toot jaayega.â€

`helper(l, r)` ka matlab hai: l se r tak ke stick ko poora cut karne ka minimum cost.
Agar `l` aur `r` ke beech koi cut possible hi nahi hai (`r - l == 1`), to cost 0 hai.
Nahi to, main har possible cut `c` try karta hoon jo `(l, r)` ke andar aata hai.
Us cut ko pehle karne ka cost hoga:

 current stick length `(r - l)`
  left part ka cost `helper(l, c)`
  right part ka cost `helper(c, r)`

Jo option sabse sasta padta hai, wahi answer.
Same `(l, r)` baar-baar na solve karna pade isliye dp mein store kar liya.

One-line yaad rakhne wali line:
ðŸ‘‰ Har range ke liye socho â€œpehla cut kaunsa hoâ€, cost = length + left + right, aur DP se repeat bachao.

'''
class Solution:
    def minCost(self, n: int, cuts: List[int]) -> int:
        # n = len(cuts)
        dp = [[-1 for _ in range(n+1)] for _ in range(n+1)]
        def helper(l,r):
            if r-l==1:return 0 # in this case we are not gonna make any cut 
            if dp[l][r]!=-1:return dp[l][r]
            cost = float("inf")
            for c in cuts:
                if l<c<r: #we can not make a cut if c is beyond the current stick
                    cost = min(cost, (r-l) + helper(c,r) + helper(l,c) )
            dp[l][r] = cost if cost!=float("inf") else 0

            return dp[l][r]
        return helper(0,n)        
Solution().minCost(n = 7, cuts = [1,3,4,5])
```

```python
# a better approach is to first sort the array first
# attach 0 and n to the endpoints of cuts
# after sorting it is guaranteed that the right half will never interfere with left half of stick
class Solution:
    def minCost(self, n: int, cuts: List[int]) -> int:
        # We sort the cuts and add the stick's natural ends (0 and n) as "virtual cuts."
        cuts.sort()
        cuts =[0] + cuts + [n]

        dp = {} #only upper trianggular matrix is needed

        def solve(i,j): #We define solve(i, j) as the minimum cost to perform all cuts between index i and index j.

            if i>j:return 0 # The Base Case: When i > j, there are no more cuts to make, so the cost is 0.
            if ( i,j) in dp:return dp[i,j]

            ans = float("inf")
            for k in range(i,j+1):
                ans = min(ans,
                          solve(i,k-1)+
                          solve(k+1,j)+
                          (cuts[j+1]-cuts[i-1])# This is the length of the current stick segment.
                          )
            dp[i,j] = ans if ans!=float("inf") else cuts[j+1]-cuts[i-1]
            return dp[i,j]
        return solve(1,len(cuts)-2 )
# o(n^3)
# o(n^2) for dp and o(n) for stack
Solution().minCost(n = 7, cuts = [1,3,4,5])
```

Burst Balloons

```python
class Solution:
    def maxCoins(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [[-1 for _ in range(n)] for _ in range(n)]
        def helper(i,j):
            if j<i:
                return 0
            if dp[i][j]!=-1:return dp[i][j]
            
            max_coins = float("-inf")
            for k in range(i,j+1):
                coins = ((nums[i-1] if i!=0 else 1)*nums[k]*(nums[j+1] if j!=n-1 else 1)) + helper(i,k-1) + helper(k+1,j)
                max_coins = max(max_coins,coins)

            dp[i][j] = max_coins
            return dp[i][j]
        return helper(0,n-1)
# o(n^3)
# o(n^2),o(n) stack space
Solution().maxCoins(nums = [3,1,5,8])
```

Evaluate Boolean Expression to True

```python
# given that alternate items of exp are boolean oerators ie "t&f|f&t"
class Solution:
    def func(self,exp):
        n = len(exp)
        dp = [[-1 for _ in range(n)] for _ in range(n)]
        def helper(i,j):
            if i==j:return (0,1) if exp[i]=="F" else (1,0)
            if i>j:return 
            if dp[i][j]!=-1:return dp[i][j]
            ways_true = 0
            ways_false = 0

            for k in range(i + 1,j,2):
                t1,f1 = helper(i,k-1)
                t2,f2 = helper(k + 1,j)
                    
                if exp[k] == "|":
                    ways_true  += t1*t2 + t1*f2 + f1*t2
                    ways_false  +=  f1*f2
                elif exp[k] == "&":
                    ways_true += t1*t2
                    ways_false   += f1*f2+ t1*f2 + f1*t2
                elif exp[k] == "^":
                    ways_true  += t1*f2 + f1*t2
                    ways_false   +=  t1*t2 + f1*f2

            dp[i][j] = ways_true,ways_false
            return dp[i][j]


        return helper(0,n-1)[0]
    
Solution().func( "T^F|F")
```

```python
class Solution:
    def func(self,exp):
        n = len(exp)
        dp = [[(0,0) for _ in range(n)] for _ in range(n)]

        for i in range(n):
            for j in range(n):
                if i==j:
                    if exp[i]=="T":dp[i][j] = (1,0)
                    elif exp[i] == "F": dp[i][j] = (0,1)

        for i in range(n-1,-1,-1): #reverse of recursion
            for j in range(0,n): #reverse of recursion
                if i>j:continue
                if i==j:continue

                ways_true = 0
                ways_false = 0
                for k in range(i+1,j,2):
                    t1,f1 = dp[i][k-1]
                    t2,f2 = dp[k + 1][j]
                        
                    if exp[k] == "|":
                        ways_true  += t1*t2 + t1*f2 + f1*t2
                        ways_false  +=  f1*f2
                    elif exp[k] == "&":
                        ways_true += t1*t2
                        ways_false   += f1*f2+ t1*f2 + f1*t2
                    elif exp[k] == "^":
                        ways_true  += t1*f2 + f1*t2
                        ways_false   +=  t1*t2 + f1*f2

                dp[i][j] = ways_true,ways_false
        return dp[0][n-1][0]
                
Solution().func( "T^F|F")
```

Palindrome Partitioning - II

```python
class Solution:
    def minCut(self, s: str) -> int:
        n = len(s)

        # 1. Pre-compute all palindromes in O(N^2)
        # is_pal[i][j] will be True if s[i:j+1] is a palindrome
        is_pal = [[False] * n for _ in range(n)]
        for length in range(1, n + 1):
            for i in range(n-length+1):
                j = i+length-1
                if s[i] == s[j]:
                    if length <= 2 or is_pal[i+1][j-1]:
                        is_pal[i][j] = True



        dp = [-1 for _ in range(n)]
        def helper(i):
            if i==n:return 0
            if dp[i]!=-1:return dp[i]
            partitions = 0 if is_pal[i][n-1] else n-i-1 #if everyone is splitted individually then all are a palindrome

            for j in range(i,n):
                if is_pal[i][j]:
                    partitions  = min(partitions,1 + helper(j+1))
            dp[i] = partitions
            return dp[i]
        return helper(0)
 
# o(n^2) for memoizatin and o(n^2) for is_pal
# o(n^2) for dp and o(n^2) for is_pal o(n) for stack 
Solution().minCut(s="leet")
```

 Partition Array for Maximum Sum

```python
class Solution:
    def maxSumAfterPartitioning(self, arr: List[int], k: int) -> int:
        n = len(arr)
        dp = [-1 for _ in range(n)]
        def helper(i):
            if i==n:return 0
            if dp[i]!=-1:return dp[i]
            s = arr[i]
            max_item = arr[i]
            for j in range(i,i+k):
                if j>n-1:continue
                max_item = max(max_item,arr[j])
                s = max(s,(j-i+1)*max_item + helper(j+1))
            dp[i] = s
            return  dp[i]
        return helper(0)
    
Solution().maxSumAfterPartitioning(arr = [1,15,7,9,2,5,10], k = 3)
```

```python
class Solution:
    def maxSumAfterPartitioning(self, arr: List[int], k: int) -> int:
        n = len(arr)
        dp = [0 for _ in range(n+1)] # if i is n then return 0

        for i in range(n-1,-1,-1):
            s = arr[i]
            max_item = arr[i]
            for j in range(i,i+k):
                if j>n-1:continue
                max_item = max(max_item,arr[j])
                s = max(s,(j-i+1)*max_item + dp[j+1])
            dp[i] = s
        return dp[0]

Solution().maxSumAfterPartitioning(arr = [1,15,7,9,2,5,10], k = 3)
```

Maximal Rectangle

```python
class Solution:
    def maximalRectangle(self, matrix: List[List[str]]) -> int:
        n = len(matrix)
        m = len(matrix[0])
        temp = [0 for _ in range(m)]
        ans = 0
        for i in range(n):
            arr = [k+int(j) if int(j)!=0 else 0 for k,j in zip(temp,matrix[i]) ]
            ans = max(ans,self.largestRectangleArea(arr))
            temp = arr[:]
        return ans

    def largestRectangleArea(self, heights: List[int]) -> int:

        def right_small(arr):
            n = len(arr)
            st = []
            next_smaller = [-1 for _ in range(n)]
            for i in range(n-1,-1,-1):
                while st and  arr[st[-1]]>=arr[i]:
                    st.pop()
                next_smaller[i] = st[-1] if st else n
                st.append(i)
            return next_smaller

        def left_small(arr):
            n = len(arr)
            st = []
            next_smaller = [-1 for _ in range(n)]
            for i in range(n):
                while st and  arr[st[-1]]>=arr[i]:
                    st.pop()
                next_smaller[i] = st[-1] if st else -1
                st.append(i)
            return next_smaller
        
        right = right_small(heights)
        left = left_small(heights)

        ans = 0
        n = len(heights)
        for i in range(n):
            ans = max(ans,(right[i] -left[i] -1)*heights[i])
        return ans

# o(n*m)
# o(m)
Solution().maximalRectangle([["0","1"],["1","0"]])
```

Count Square Submatrices with All Ones

```python
class Solution:
    def countSquares(self, matrix: List[List[int]]) -> int:
        n,m = len(matrix),len(matrix[0])
        dp = [[matrix[i][j] if i == 0 or j == 0 else 0
                for j in range(m)]
                for i in range(n)]
        for i in range(1,n):
            for j in range(1,m):
                dp[i][j] = (1 + min(dp[i-1][j],dp[i-1][j-1],dp[i][j-1])) if matrix[i][j]==1 else 0

        return sum([sum(l) for l in dp])

Solution().countSquares( matrix =
[
  [0,1,1,1],
  [1,1,1,1],
  [0,1,1,1]
])

```
