Best Time to Buy and Sell Stock I

```python
class Solution:
    def buySell(self,arr):
        n = len(arr)
        mini = arr[0]
        ans = 0
        for i in range(1,n):
            profit = arr[i]-mini
            ans = max(ans,profit)
            mini = min(mini,arr[i])
        return ans
Solution().buySell([7,1,5,3,6,4])
```

Best Time to Buy and Sell Stock II

```python
# in top down -> Profit calculate tabhi hota hai jab recursion 'hit' karke wapas aa raha hota hai.
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        n = len(prices)
        """ 
        0:can't buy
        1:free to buy
        """
        dp = [[-1 for _ in range(2)] for _ in range(n)]
        def helper(i,buy):
            if i==n:
                return 0
            if dp[i][buy]!=-1:return dp[i][buy]

            if buy:
                profit = max(helper(i+1,0) - prices[i], helper(i+1,1))
            else:#sell
                profit = max(helper(i+1,1) + prices[i], helper(i+1,0))

            dp[i][buy] = profit
            return dp[i][buy]

            
        return helper(0,1)
    
Solution().maxProfit([7,1,5,3,6,4])
```

Best Time to Buy and Sell Stock with Cooldown

```python
# in top down -> Profit calculate tabhi hota hai jab recursion 'hit' karke wapas aa raha hota hai.
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        n = len(prices)
        """ 
        0:can't buy and can't sell because sold on prev day
        1:free to buy
        2:free to sell
        """
        dp = [[-1  for _ in range(3)] for _ in range(n)]

        def helper(i:int,buy:int)->int:
            # Base Case 1: Din khatam
            if i==n:
                return 0

            if dp[i][buy]!=-1:return dp[i][buy]

            if buy==0:
                profit = helper(i+1,1) #sending 1 to next state so that it can buy
            elif buy==1:
                profit = max(helper(i+1,2) - prices[i], helper(i+1,1))
                # if i buy then i will tell next state to cooldown so send 0 to signal that
            elif buy==2:
                profit = max(helper(i+1,0)+ prices[i], helper(i+1,2))

            dp[i][buy] = profit

            return dp[i][buy]

            
        return helper(0,1)
    
Solution().maxProfit( prices = [1,2,3,0,2]) 
```

Buy and Sell Stocks With Transaction Fee

```python
# in top down -> Profit calculate tabhi hota hai jab recursion 'hit' karke wapas aa raha hota hai.
class Solution:
    def maxProfit(self, prices: List[int],fee:int) -> int:
        n = len(prices)
        """ 
        0:can't buy
        1:free to buy
        """
        dp = [[-1 for _ in range(2)] for _ in range(n)]
        def helper(i,buy):
            if i==n:
                return 0
            if dp[i][buy]!=-1:return dp[i][buy]

            if buy:
                profit = max(helper(i+1,0) - prices[i], helper(i+1,1))
            else:#sell
                profit = max(helper(i+1,1) + prices[i]-fee, helper(i+1,0))

            dp[i][buy] = profit
            return dp[i][buy]

            
        return helper(0,1)
    
Solution().maxProfit(prices = [1,3,2,8,4,9], fee = 2)
```
