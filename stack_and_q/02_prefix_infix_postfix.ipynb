Infix to postfix

```python
class Solution:
    def infix_to_postfix(self,exp):
        
        st = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}
        ans = []
        for char in exp:
            if char.isalnum():ans.append(char)
            elif char=="(":st.append(char)
            elif char==")":
                while st and st[-1]!="(":
                    ans.append(st.pop())
                st.pop()
            else:
                while st and st[-1]!="(" and operators[char]<=operators[st[-1]]:
                    if char=="^" and operators[char]==operators[st[-1]]:break#strictly greater not required for ^ operator so we dont pop
                    ans.append(st.pop())
                st.append(char)
                
        while st:
            ans.append(st.pop())

        return "".join(ans)
# o(n)
# o(n)   
Solution().infix_to_postfix("f+d-c*(b+a)")
```

Infix to Prefix

```python
class Solution:
    def infix_to_prefix(self,exp:str)->str:
        st = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}
        ans = []
        rev_exp = [")" if i=="(" else "(" if i==")" else i for i in exp[::-1] ]

        for char in rev_exp:
            if char.isalnum():ans.append(char)
            elif char=="(":st.append(char)
            elif char==")":
                while st and st[-1]!="(":
                    ans.append(st.pop())
                st.pop()
            else:
                while st and st[-1]!="(" :
                    top_prec = operators[st[-1]]
                    curr_prec = operators[char]
                    # Pop if top is stronger
                    # OR if they are equal AND char is '^' (Right-associative mirrored)
                    if top_prec > curr_prec or (top_prec == curr_prec and char == "^"):
                        ans.append(st.pop())
                    else:
                        break
                st.append(char)
                
        while st:
            ans.append(st.pop())

        return "".join(ans[::-1])
# o(n)
# o(n) 
Solution().infix_to_prefix("(a+b)*c-d+f")
```

Postfix to Infix

```python
class Solution:
    def postfix_to_infix(self,exp:str):
        st = []
        ans = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}

        for char in exp:
            if char.isalnum():st.append(char)
            elif char in operators:
                operand1 = st.pop()
                operand2 = st.pop()
                st.append("".join(["(",operand2,char,operand1,")"]))#this line alone might take o(n) time because of adding stings
        return st[0]
# o(n)
# o(n)
Solution().postfix_to_infix("ab-de+f*/")    
```

Prefix to Infix

```python
class Solution:
    def prefix_to_infix(self,exp:str):
        st = []
        ans = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}

        for char in exp[::-1]:
            if char.isalnum():st.append(char)
            elif char in operators:
                operand1 = st.pop()
                operand2 = st.pop()
                st.append(["(",operand1,char,operand2,")"])#this line alone might take o(n) time because of adding stings thats why we are using list

        # The final step: A single join at the very end
        # This is where the linear O(N) work happens
        def flatten(nested):
            result = []
            def _flat(item):
                if isinstance(item,list):
                    for sub in item:
                        _flat(sub)
                else:
                    result.append(item)
            _flat(nested)
            return "".join(result)

        return flatten(st[0])
# o(n)
# o(n)
Solution().prefix_to_infix("*+pq-mn")    
```

Postfix to Prefix

```python
class Solution:
    def postfix_to_prefix(self,exp:str):
        st = []
        ans = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}

        for char in exp:
            if char.isalnum():st.append(char)
            elif char in operators:
                operand1 = st.pop()
                operand2 = st.pop()
                st.append([char,operand2,operand1])

        def flatten(nested):
            result = []
            def _flat(item):
                if isinstance(item,list):
                    for i in item:
                        _flat(i)
                else:
                    result.append(item)
            _flat(nested)
            return "".join(result)
        
        return flatten(st[0])
    
# o(n)
# o(n)
Solution().postfix_to_prefix("ab-de+f*/")    
```

Prefix to Postfix

```python
class Solution:
    def prefix_to_postfix(self,exp:str):
        st = []
        ans = []
        operators = {"^": 3, "/": 2, "*": 2, "-": 1, "+": 1}

        for char in exp[::-1]:
            if char.isalnum():st.append(char)
            elif char in operators:
                operand1 = st.pop()
                operand2 = st.pop()
                st.append([operand1,operand2,char])

        def flatten(nested):
            result = []
            def _flat(item):
                if isinstance(item,list):
                    for i in item:
                        _flat(i)
                else:
                    result.append(item)
            _flat(nested)
            return "".join(result)
        
        return flatten(st[0])
    
# o(n)
# o(n)
Solution().prefix_to_postfix('/-ab*+def')    
```
