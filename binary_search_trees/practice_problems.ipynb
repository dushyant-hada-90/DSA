```python
# Definition for a binary tree node.

from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree_from_level_order(arr):
    if not arr or arr[0] is None:
        return None

    root = TreeNode(arr[0])
    q = deque([root])
    i = 1

    while q and i < len(arr):
        node = q.popleft()

        # Left child
        if i < len(arr) and arr[i] is not None:
            node.left = TreeNode(arr[i])
            q.append(node.left)
        i += 1

        # Right child
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            q.append(node.right)
        i += 1

    return root

```

Ceil in a Binary Search Tree

```python
class Solution:
   
    def ceil(self,root,key):
        ceil_val = float("inf")

        while root:
            if root.val == key:
                return root.val
            
            elif root.val<key:
                root = root.right
            elif root.val>key:
                ceil_val = root.val
                root = root.left
        return ceil_val if ceil_val != float("inf") else -1

sol = Solution()
root = build_tree_from_level_order([10,5,13,3,6,11,14,2,4,9,None,None])
sol.ceil(root,3.5)
```

Floor in a Binary Search Tree

```python
class Solution:
    def floor(self,root,key):
        floor_val = float("-inf")

        while root:
            if root.val == key:
                return root.val
            elif root.val>key:
                root = root.left
            elif root.val<key:
                floor_val = root.val
                root = root.right

        return floor_val if floor_val != float("-inf") else -1
# o(logn) ,o(1)
sol = Solution()
sol.floor(root,3.5)
```

Insert a given Node in Binary Search Tree


```python
class Solution:
    def insertIntoBST(self, root, val):
        # print(root.val)
        if not root:
            return TreeNode(val)

        if root.val>val:
            root.left = self.insertIntoBST(root.left,val)
            
        elif root.val<val:
            self.insertIntoBST(root.right,val)
    
        return root
    
# o(h),o(h) stack space
sol = Solution()
root = build_tree_from_level_order([4,2,7,1,3,5])
sol.insertIntoBST(root,9).val
```

Delete Node in a BST

```python
class Solution:
    def deleteNode(self, root, key):
        if not root:
            return
            
        if root.val>key:
            root.left = self.deleteNode(root.left,key)
        elif root.val<key:
            root.right = self.deleteNode(root.right,key)

        else: # node found
             # Case 1: No child
            if not root.right and not root.left:
                return None
            # Case 2: One child
            if not root.left:
                return root.right
            if not root.right:
                return root.left
             # Case 3: Two children â†’ replace with  inorder_successor (we can also use inorder predecessor)
            inorder_successor = self.get_is(root.right)
            root.val = inorder_successor.val
            root.right = self.deleteNode(root.right,inorder_successor.val)
        
        return root
    
    def get_is(self,node):
        while node.left:
            node = node.left
        return node
            
```

Kth Smallest Element in a BST

```python
class Solution:
    def kthSmallest(self, root, k):
        if not root:
            return
        
        # morris inorder traversal
        curr = root
        while curr:
           
            if curr.left == None:
                k-=1
                if k==0:
                    return curr.val
                curr = curr.right
            else:
                ip = curr.left
                while (ip.right) and (ip.right != curr):
                    ip = ip.right
                if ip.right == None:
                    ip.right = curr
                    curr = curr.left
                else:
                    ip.right = None
                    k-=1
                    if k==0:
                        return curr.val
                    curr = curr.right        
```

Check if a tree is a BST or BT

```python
class Solution:
    def isValidBST(self, root):
        
        def helper(node,low,high):
            if not node:
                return True
            if not (low < node.val <high):
                return False
            return helper(node.left,low,node.val) and helper(node.right,node.val,high)
        
        return helper(root,float("-inf"),float("inf"))
            
sol = Solution()
sol.isValidBST(root)
        
```

LCA in Binary Search Tree

```python
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        if not root:
            return
        
        if root.val > p.val and root.val > q.val:
            return self.lowestCommonAncestor(root.left,p,q)
        if root.val < p.val and root.val < q.val:
            return self.lowestCommonAncestor(root.right,p,q)
        else: #ek root ke aage hai aur ek peeche ie root hi lca hai
            return root
# o(h),o(h)stack space
```

Construct a BST from a preorder traversal

```python
# a brute force approch could be to create root from the 0th element of preorder and then pass remaining items one by one to a function insertIntoBst 
# time complexity for this would be o(n**2)
```

```python
class Solution:
    def bstFromPreorder(self, preorder):
        self.i = 0
        def build(ub = float("inf")):
            if (self.i == len(preorder))  or (preorder[self.i] > ub):
                return
            
            root = TreeNode(preorder[self.i])
            self.i+=1

            root.left = build(root.val)
            root.right = build(ub)

            return root

        return build()
            
                       
sol = Solution()
sol.bstFromPreorder([8,5,1,7,10])
```

Inorder Successor/Predecessor in BST

```python
class Solution:
    def inorderSuccessor(self,root,target):
        succ = target
        def helper(root):
            nonlocal succ,target
            if not root:
                return
            # print(root.val)
            if root.val <= target:
                helper(root.right)
            elif root.val > target:
                succ = root.val
                helper(root.left)
        helper(root)
        return succ if succ != target else -1
        

    def inorderPredecessor(self,root,target):
        pred = target
        def helper(root):
            nonlocal pred,target
            if not root:
                return
            # print(root.val)
            if root.val < target:
                pred = root.val
                helper(root.right)
            elif root.val >= target:
                helper(root.left)
        helper(root)
        return pred if pred != target else -1

sol = Solution()
root = build_tree_from_level_order([5,3,7,2,4,6,9,1,None,None,None,None,None,8,10])
sol.inorderSuccessor(root,5),sol.inorderPredecessor(root,5)
```

BST Iterator

```python
class BSTIterator:

    def __init__(self, root):
        self.stack = []
        curr = root
        # initially push all leftist elements into the stack
        while curr:
            self.stack.append(curr)
            curr = curr.left
        

    def next(self) -> int:
        node =  self.stack.pop()
        curr = node.right
        # go to right node and the push all leftist elements of it into stack
        while curr:
            self.stack.append(curr)
            curr = curr.left
        return node.val

        

    def hasNext(self) -> bool:
        return bool(self.stack) #if stack is empty then return False else True
        

# o(1),i am pushing n elements into the stack and there will be n calls to print all elements so on an average time comp is o(1)

# o(h) i am only storing height of tree and not even complete tree height so even in worst case of skewed tree the space comp. is o(h)
```

Two Sum In BST | Check if there exists a pair with Sum K

```python
class Iterator:
    # the aim is to create 2 iterators that could return next and prev. nodes
    # if next can generate inorder in ascending order then prev can generate inorder in descending order
    def __init__(self, root):
        self.stack_next = []
        self.stack_prev = []
        # initially push all leftist elements into the stack
        curr = root
        while curr:
            self.stack_next.append(curr)
            curr = curr.left
        # initially push all rightist elements into the stack
        curr = root
        while curr:
            self.stack_prev.append(curr)
            curr = curr.right
        
    def next(self) -> int:
        node =  self.stack_next.pop()
        curr = node.right
        # go to right node and the push all leftist elements of it into stack
        while curr:
            self.stack_next.append(curr)
            curr = curr.left
        return node.val
    
    def prev(self) -> int:
        node =  self.stack_prev.pop()
        curr = node.left
        # go to left node and the push all rightist elements of it into stack
        while curr:
            self.stack_prev.append(curr)
            curr = curr.right
        return node.val
    

class Solution:
        

    def findTarget(self, root, k):
        self.iterator = Iterator(root) #create an iterator object using root to access next and prev
        self.small,self.large = self.iterator.next(),self.iterator.prev() #initially set both pointers to two edges
        while self.large>self.small:  # run only till large is more than small , if the ever cross each other or arriv at same midddle node then break
            print(self.small,self.large)
            if self.small+self.large == k:
                return True
            if self.small+self.large<k:
                self.small = self.iterator.next() #if sum is less than move the small pointer to increase its value
            if self.small+self.large>k:
                self.large = self.iterator.prev() #if sum is larger then move the large pointer to decrease its value
        return False
    
# o(n) worst case when we might need to visit each node to check sum
# o(2h) for storing inorder and reverse inprder traversals which won't be more than h of tree because we are not storing entire traversal simulataneously 

sol = Solution()
root=build_tree_from_level_order([7,3,10,2,6,9,11,1,None,5,None,8,None,None,None,None,None,4,None])
sol.findTarget(root,10)
```

Largest BST in Binary Tree

```python
# brute force approach could be to traverse through the tree and pass each node to a helper function - isBST(self,root) , if isBST returns True then pass that root to another helper function countNodes(root) and keep a variable tht store root of bst with max_nodes

# since we are calling isBST for each node and isBST inherently visits each node in worst case so time comp. of isBST is o(n^2) , the time complexity of countNodes is o(n) so overall i is o(n^2) operation
```

```python
class Solution:
    def largestBst(self,root):

        def helper(root):
            if not root:
                return 0,float("-inf"),float("inf") # return count_of_nodes in sthat subtree , max value , min value bounds
            
            l_count,l_max,l_min = helper(root.left)
            r_count,r_max,r_min = helper(root.right)

            if l_max<root.val<r_min: # to be a valid bst, my current node val should be less than the smallest node of right subtree and larger than the largest node of left subtree
                return (1 + l_count + r_count),l_min,r_max #return count,(abhi thak tumne sabse bada kya dekha hai),(abhi thak tumne sabse chota kya dekha hai)
            else:
                return (1 + l_count + r_count),float("inf"),float("-inf") # return such high values in max and such low values in min so that the if condition is never true in subsequent recursions , because logically if a subtree is not a valid bst then its parent , parent'parent,parent'sparent'parent,... can't be a valid bst so the condition had to be made false all along the way
    
        return helper(root)[0]
sol = Solution()
root = build_tree_from_level_order([20,15,40,14,18,30,60,None,17,16,19,None,None,None,None])
sol.largestBst(root)
        
```
