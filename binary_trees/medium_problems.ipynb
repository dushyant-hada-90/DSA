 Height of Binary Tree 

```python
class Solution:
    def maxDepth(self, root) :
        if not root:
            return 0
        left = (root.left and self.maxDepth(root.left)) or 0
        right = (root.right and self.maxDepth(root.right)) or 0
        h= 1 + max(left , right)

        return h
# o(n) n=nuber of nodes  , o(h) where h=height of tree
```

 Check for Balanced Binary Tree

```python
class Solution:
    def balanced(self,root):  #o(n)

        def find_height(root): #o(n)
            if root == None:
                return 0
            left = find_height(root.left)
            right = find_height(root.right)
            height = 1+max(left,right)
            return height
        
        if root == None:
                return True
        
        left_height = find_height(root.left)  
        right_height = find_height(root.right)

        if abs(left_height-right_height)>1:
            return False
        else:
            right_bal = self.balanced(root.right)
            left_bal = self.balanced(root.left)
        return right_bal and left_bal

# o(n**2) ,o(h) 
```

```python
sol = Solution()
sol.balanced(root)
```

```python
class Solution:
    def balanced(self,root):  #o(n)

        def dfs(root): #o(n)
            if root == None:
                return 0
            left = dfs(root.left)
            if left == -1:
                 return -1
            right = dfs(root.right)
            if right == -1:
                 return -1

            if abs(left-right)>1:
                return -1
            
            height = 1+max(left,right)
            return height
        
        return dfs(root) != -1

# o(n) ,o(n) 
```

Diameter of Binary Tree

```python
class Solution:
    max_h = 0
    def diameter(self,root):
        if not root:
            return 
        
        def find_height(root): #o(n)
            if root == None:
                return 0
            left = find_height(root.left)
            right = find_height(root.right)
            height = 1+max(left,right)
            return height
        
        lh = find_height(root.left)
        rh = find_height(root.right)

        self.max_h = max(self.max_h,lh+rh)

        self.diameter(root.left)
        self.diameter(root.right)

        return self.max_h

# o(n**2),o(h)
```

```python
sol = Solution()
sol.diameter(root)
```

Maximum path sum

```python
class Solution:
   
    def maxPathSum(self, root) -> int:
        ans = float("-inf")
        def helper(root):
            if not root:
                return 0
            nonlocal ans
            print(root.val,ans,"root.val")
            l_sum = max(helper(root.left), 0)
            r_sum = max(helper(root.right), 0)

            curr_sum = max(root.val,l_sum+root.val, r_sum+root.val )
            ans = max(ans,curr_sum,root.val + l_sum  + r_sum )
            print(ans,"ans")
            return curr_sum

        helper(root)
        return ans
# o(n)
sol = Solution()
sol.maxPathSum(root)
```

Same Tree

```python
root1 = Node(1)
root1.right = Node(3)
root1.left = Node(2)

root2 = Node(1)
root2.right = Node(3)
root2.left = Node(2)

```

```python
class Solution:
    def isSameTree(self, p , q) -> bool:
       
        if p == q ==None:
            return True
        if p==None or q==None or p.val!=q.val:
            return False

        left = self.isSameTree(p.left,q.left)
        right = self.isSameTree(p.right,q.right)
        return left and right 
            
# o(n),O(h)
sol = Solution()
sol.isSameTree(root2,root1)
```

Zig Zag Traversal of Binary Tree

```python
class Solution:
    def zigzagLevelOrder(self, root) :
        if not root:
            return []
        q = [root]
        ans=[]
        direction = 1
        
        while q :
            level = []
            next_level = []
            for node in q:
                level.append(node.val)
                node.left and next_level.append(node.left)
                node.right and next_level.append(node.right)
            
            ans.append(level[::direction])
            direction *= -1
            q= next_level
    
        return ans
# o(n) ,o(w) as stack space and o(n) for storing ans
```

Boundary Traversal

```python
class Solution:
    def boundary_traversal(self,root):

        if not root:
            return []
        
        result = [root.val] if not self.is_leaf(root) else []

        self.add_left_boundary(root.left,result)
        self.add_leaves(root,result)

        right_boundary = []
        self.add_right_boundary(root.right,right_boundary)
        result+=right_boundary[::-1]
    
        return result
        
    def is_leaf(self,node):
        return (node) and (not node.left) and (not node.right)
    
    def add_left_boundary(self,node,result):
        while node:
            if not self.is_leaf(node):
                result.append(node.val)
            if node.left:
                node=node.left
            else:
                node = node.right
    def add_right_boundary(self,node,result):
        while node:
            if not self.is_leaf(node):
                result.append(node.val)
            if node.right:
                node = node.right
            else:
                node = node.left

    def add_leaves(self,node,result): #traverse each node
        if node:
            if self.is_leaf(node):
                result.append(node.val)
                return
            self.add_leaves(node.left,result)
            self.add_leaves(node.right,result)



sol = Solution()
sol.boundary_traversal(root)
```

Vertical Order Traversal of Binary Tree 

```python
from collections import defaultdict, deque

class Solution:
    def verticalTraversal(self, root) :
        if not root:
            return []
        
        nodes = []
        q = deque([(root, 0, 0)])
        
        while q :      # BFS -> O(n)
            node, row, col = q.popleft()
            nodes.append((col, row, node.val))
            node.left and q.append((node.left,row+1,col-1))
            node.right and q.append((node.right,row+1,col+1))
        nodes.sort()  # sorting-> o(nlogn)
        print(nodes)

        res = defaultdict(list)
        for col,row,val in nodes:  #O(n)
            res[col].append(val)
        print("res->",res)
        
        return [res[x] for x in sorted(res.keys())]

sol = Solution()
sol.verticalTraversal(root)

# O(nlogn) , O(n) suxiliary space, o(w) stack space where w is width of tree
```

 Top View of Binary Tree

```python
# from collections import defaultdict

class Solution:
    def top_view(self,root):
        if not root:
            return []
        nodes = []
        def dfs(root,row,col):  # O(n)
            if not root:
                return
            nodes.append((col,row,root.val))
            dfs(root.left,row+1,col-1)
            dfs(root.right,row+1,col+1)
        dfs(root,0,0)
        nodes.sort() # O(nlogn)
        print(nodes)

        top_nodes = {}
        for  col,row,val in nodes:  # O(n)
            if col not in top_nodes:
                top_nodes[col] =  val
        return [top_nodes[i] for i in sorted(top_nodes.keys())]  # O(nlogn)

sol = Solution()
sol.top_view(root)

# O(nlogn) , O(n) auxiliary space and o(h) stack space
```

Bottom View of Binary Tree

```python
from collections import deque

class Solution:
    def bottom_view(self,root):
        if not root:
            return []
        bottom_nodes = {}
        q = deque([(root, 0)]) 

        while q:  #o(n)
            node, col = q.popleft()
            # update bottom-most node for this column
            bottom_nodes[col] = node.val

            if node.left:
                q.append((node.left, col - 1))
            if node.right:
                q.append((node.right, col + 1))

        
        print(bottom_nodes)
        return [bottom_nodes[i] for i in sorted(bottom_nodes.keys())]  # O(nlogn) in worst case skewed tree only

sol = Solution()
sol.bottom_view(root)

# O(nlogn) , O(n) auxiliary space and o(w) stack space
```

Right/Left View of Binary Tree

```python
class Solution:
    def rightSideView(self, root):
        if not root:
            return []
        nodes = []
        def dfs(root,row): # O(n)
            if not root:
                return
            if len(nodes) == row:
                nodes.append(root.val)
            dfs(root.right,row+1)
            dfs(root.left,row+1)
        dfs(root,0)  
        
        return nodes

sol = Solution()
sol.rightSideView(root)
            
# O(n),O(H) stack space
```

Symmetric Binary Tree

```python
from collections import deque

class Solution:
    def isSymmetric(self, root):
        if not root:
            return True
        level_nodes = []
        q = deque([(root.left,root.right)])

        while q:
            t1,t2 = q.popleft()
            
            if not t1 and not t2:
                continue
            if not t1 or not t2:
                return False
            if t1.val != t2.val:
                return False
            
            q.append((t1.left,t2.right))
            q.append((t1.right,t2.left))
        return True
        

sol =Solution()
sol.isSymmetric(root)
# o(n) ,o(w) sytack sspace and auxiliary space is o(n)
```
