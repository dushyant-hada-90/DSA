```python
# Definition for a binary tree node.

from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree_from_level_order(arr):
    if not arr or arr[0] is None:
        return None

    root = TreeNode(arr[0])
    q = deque([root])
    i = 1

    while q and i < len(arr):
        node = q.popleft()

        # Left child
        if i < len(arr) and arr[i] is not None:
            node.left = TreeNode(arr[i])
            q.append(node.left)
        i += 1

        # Right child
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            q.append(node.right)
        i += 1

    return root

```

LCA in Binary Tree

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        
        def dfs(node):
            if not node:
                return None
        
            if node==p or node==q:
                return node
                
            left =  dfs(node.left) 
            right = dfs(node.right)

            if left and right:
                return node
            
            return left or right
            
        return dfs(root)

# o(n),o(h)
        
```

Maximum width of a Binary Tree

```python
from collections import deque
# we assign each node index and then find diff of first and last index of each node and do +1 to find width at that level
class Solution:
    def widthOfBinaryTree(self, root) -> int:

        if not root:
            return 0
        
        q=deque([(1,root)])
        max_width = 0

        while q:
            level_length = len(q)
            first_index = q[0][0]
            last_index = q[-1][0]

            max_width = max(max_width,last_index - first_index+1)
            
            for _ in range(level_length):
                index,node = q.popleft()
                node.left and q.append((2*index , node.left))
                node.right and q.append((2*index+1 , node.right))

        return max_width

# o(n) , o(n) for deque
sol = Solution()
sol.widthOfBinaryTree(build_tree_from_level_order([1,3,2,5,None,None,9,6,None,7]))
```

Children Sum Property enforcement

```python
class Solution:
    def childrenSumProperty(self,root):

        def helper(node):
            if not root or (not node.left and not node.right):
                return
            
            helper(node.left)
            helper(node.right)

            left_val = node.left.val if node.left else 0
            right_val = node.right.val if node.right else 0
            child_sum = left_val + right_val

            if child_sum >= node.val:
                node.val = child_sum
            else:
                increment(node,node.val-child_sum)
        
        def increment(node,diff):
            if node.left:
                node.left.val += diff
                helper(node.left)
            elif node.right:
                node.right.val += diff
                helper(node.right)
                
        helper(root)
        return root
```

Print all the Nodes at a distance of K in a Binary Tree

```python
from collections import deque
class Solution:
    def distanceK(self, root ,target,k):
        if not root:
            return []
        q  = deque([root])
        level = []
        d = {root:None}
        while q:
            node = q.popleft()                
            if node.left:
                q.append(node.left)
                d[node.left] = node
            if node.right:
                q.append(node.right)
                d[node.right] = node
        # print(d)


        visited = set()
        ans = []
        def search(node,dist=0):

            if not node:
                return
            # print(node.val,[i.val for i in visited])   

            if node in visited:
                return
            
            if dist==k:
                ans.append(node.val)
                return
            
            visited.add(node)

            parent = d[node]
            parent and search(parent,dist+1)

            node.left and search(node.left,dist+1)

            node.right and search(node.right,dist+1)
            

        search(target)

        return ans

            
# o(n) , o(n)

sol = Solution()
root = build_tree_from_level_order([3,5,1,6,2,0,8,None,None,7,4,None,None,None,None])
sol.distanceK(root,root.left,2)
```

Minimum time taken to BURN the Binary Tree from a Node

```python
from collections import deque
class Solution:
    def timeToBurn(self,root,target):
        if not root:
            return 0
        q = deque([root])
        parent = {root:None}
        while q:
            node = q.popleft()
            if node.left:
                q.append(node.left)
                parent[node.left] = node
            if node.right:
                q.append(node.right)
                parent[node.right] = node
        
        visited,ans = set() , 0
        def visit(node,time):
            nonlocal ans
            # print(ans,node.val)
            if (not node) or (node in visited):
                return
            visited.add(node)
            ans = max(ans,time)
            
            visit(node.left,time+1)
            visit(node.right,time+1)
            visit(parent[node],time+1)
        visit(target,0)
        return ans

            

            
# o(n) , o(n) auxiliary space
sol = Solution()
root = build_tree_from_level_order([1,2,3,4,None,5,6,None,7,None,None,None,None])
sol.timeToBurn(root,root.left)
```

Count total Nodes in a COMPLETE Binary Tree

```python
class Solution:
    def countNodes(self, root) -> int:  
        
        node,left_height = root,0
        while node:  #logn
            left_height+=1
            node = node.left

        node,right_height = root,0
        while node:  #logn
            right_height+=1
            node = node.right

        if left_height==right_height:
            return 2**left_height-1
        else:
            return (1+self.countNodes(root.left) + self.countNodes(root.right))

# o((logn)**2) , o(h) stack space
sol = Solution()
root = build_tree_from_level_order([1,2,3,4,5,6,7,8,9,19,11])
sol.countNodes(root)
```

Construct the Binary Tree from Postorder and Inorder Traversal

```python
class Solution:
    def buildTree(self, inorder, postorder):
        
        hashmap = {item:index for index,item in enumerate(inorder)}

        print(hashmap)

        def helper(left,right):

            if left>right or not postorder:
                return
            
            root = TreeNode(postorder.pop())
            mid = hashmap[root.val]

            root.right = helper(mid+1,right)
            root.left = helper(left , mid-1)

            return root
        
        return helper(0,len(inorder)-1)

# o(n) , o(n) stack space ,o(n) auxiliary space
sol = Solution()
sol.buildTree(inorder = [9,3,15,20,7],postorder = [3,9,20,15,7]) 

                
        
```

 Serialize and De-serialize Binary Tree

```python
from collections import deque
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        if not root:
            return ""
        
        q=deque([root])
        level = []
        while q:
            node = q.popleft()
            if node:
                level.append(str(node.val))
                q.append(node.left)
                q.append(node.right)
            else:
                level+="#"
        return ",".join(level)
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        if not data:
            return None
        
        values = data.split(",")
        root = TreeNode(int(values[0]))
        q = deque([root])
        i = 1
        while q:
            node = q.popleft()
            if values[i] != "#":
                node.left = TreeNode(int(values[i]))
                q.append(node.left)
            i+=1

            if values[i] != "#":
                node.right = TreeNode(int(values[i]))
                q.append(node.right)
            i+=1
        
        return root
            


        
sol = Codec()
root = build_tree_from_level_order([1,2,3])
coded = sol.serialize(root)
decoded = sol.deserialize(coded)
print(coded,decoded,root)
```

Morris Preorder Traversal of a Binary Tree

```python
class Solution:
    def preorderTraversal(self, root):
        curr = root
        ans = []
        while curr:
            if curr.left == None:
                ans.append(curr.val)
                curr = curr.right
            else:
                ip = curr.left
                while (ip.right != None) and (ip.right != curr):
                    ip = ip.right

                if ip.right == None:
                    ans.append(curr.val)
                    ip.right = curr
                    curr = curr.left
                else:
                    ip.right = None
                    curr = curr.right
        return ans

# o(2n) , o(1)
sol = Solution()
sol.preorderTraversal(root)
```

Flatten Binary Tree to Linked List

```python
class Solution:
    next_right = None
    def flatten(self, root) -> None:
        
        if root == None:
            return
        
        self.flatten(root.right)
        self.flatten(root.left)

        root.left = None
        root.right = self.next_right
        self.next_right = root
# o(n) ,o(1)
sol = Solution()
sol.flatten(root)
        
```
