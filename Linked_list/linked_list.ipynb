Insertion after a node in LL

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self,value):
        # create a new node with given value
        new_node = Node(value)
        # point the address pointer of new node towards current head
        new_node.next = self.head
        # reassign head to new node
        self.head = new_node

    def traverse(self):
        # define a temporary iterator temp
        temp = self.head
        # loop through entire array till you find an item whose .next is null i.e the last item of ll
        while temp != None:
            print(temp.data,"->",end=" ")
            temp=temp.next

    def insert_after(self,prev_data,new_data):
        prev = self.head
        while prev is not None:
            if prev.data == prev_data:
                new_node = Node(new_data)
                new_node.next = prev.next
                prev.next = new_node
                break
            prev = prev.next
        if prev is None :
            return f"{prev_data} not in ll"
```
