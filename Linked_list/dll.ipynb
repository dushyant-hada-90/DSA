```python
# Definition of doubly linked list:
class ListNode:
    def __init__(self, val=0, next=None, prev=None):
        self.val = val
        self.next = next
        self.prev = prev
```

```python
class Solution:
    def insertBeforeHead(self, head, X):
        new_head = ListNode(X)
        new_head.next = head
        if head:
            head.prev = new_head
```

```python
def reverseDLL(self, head):
    current = head
    while current:
        current.next,current.prev = current.prev,current.next
        prev_node = current # so that we can return safely
        current = current.prev
    return prev_node
```

Delete all occurrences of a key in DLL

```python
def deleteAllOccurrences(self, head, target):
    if not head:
        return head

    
    current = head
    while current:
        next_node=current.next
        if current.val == target and current == head: #first node
            head=next_node
            if head:
                head.prev=None
        if current.val == target and current.next:   #middle node
            current.prev.next = current.next
            current.next.prev = current.prev

        if current.val == target and not current.next: #last node
            current.prev.next = None

        # now unlink current node
        current.next = current.prev = None
        current = next_node
    return head
# o(n),o(1)
```

Remove duplicates from sorted DLL

```python
def removeDuplicates(self, head):
    if not head:
        return head
    temp =  head
    
    while temp and temp.next:
        if temp.val == temp.next.val:
            temp.next = temp.next.next
            if temp.next:
                temp.next.prev = temp
        else:
            temp = temp.next


    return head
# o(n),o(1)
    
```

Find pairs with given sum in DLL(sorted)

```python
def findPairsWithGivenSum(self, head, target):
    temp1 = head
    l = []
    while (temp1):
        temp2 = temp1.next
        while temp2 and  temp1.val+temp2.val<=target:
            if temp1.val+temp2.val==target:
                l.append([temp1.val,temp2.val])
            temp2 = temp2.next
        temp1 = temp1.next
    return l
# o(n**2),o(k) where k is number of pairs, worst case n**2
```

```python
def findPairsWithGivenSum(self, head, target):
    l = r = head
    ds = []
    while r.next:
        r=r.next
    
    while l!=r and  r.next!=l :
        s = l.val+r.val
        if s>target:
            r=r.prev
        elif s<target:
            l=l.next
        elif s==target:
            ds.append([l.val,r.val])
            r,l=r.prev,l.next
    return ds

# o(n),o(k) where k is number of pairs, worst case n**2
```
